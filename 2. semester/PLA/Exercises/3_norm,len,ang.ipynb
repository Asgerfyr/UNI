## Exercise 1
Define the following vectors of $\mathbb{R}^3$

\begin{align*}
\mathbf{u}=(1,5,7)\\
\mathbf{v}=(2,5,9)
\end{align*}

#### Task 1
Calculate the three standard lengths of the vectors in numpy

That is $\lVert \cdot \rVert_{\infty}$,  $\lVert \cdot \rVert_{1}$ and  $\lVert \cdot \rVert_{2}$

#### Task 2
Calculate the distance between $\mathbf{u}$ and $\mathbf{v}$ in the metrics induced by the three norms in numpy

#### Task 3
Verify that the triangle inequality $\lVert \mathbf{u}+\mathbf{v}\rVert \leq \lVert \mathbf{u}\rVert + \lVert \mathbf{v}\rVert$
in all three norms

```python
import numpy as np

u = np.array([1,5,7])
v = np.array([2,5,9])

norm_type = [np.inf, 1, 2]
norm_names = ['max', 'l1', 'l2']

#task 1
print("task 1")

u_norms = []
v_norms = []

for i in range(len(norm_type)):
    u_norms.append(np.linalg.norm(u, ord=norm_type[i]))
    v_norms.append(np.linalg.norm(v, ord=norm_type[i]))
    print(f'norm {norm_names[i]} of u: {u_norms[i]} ')
    print(f'norm {norm_names[i]} of v: {v_norms[i]} ')

print("\n\n")
#task 2
print("task 2")

dif = u-v

dif_norms = []

for i in range(len(norm_type)):
    dif_norms.append(np.linalg.norm(dif, ord=norm_type[i]))
    print(f'norm {norm_names[i]} of the difference: {dif_norms[i]} ')



print("\n\n")
#task 3
print("task 3")
sum = u+v

sum_norms = []

for i in range(len(norm_type)):
    sum_norms.append(np.linalg.norm(sum, ord=norm_type[i]))
    print(f'norm {norm_names[i]} of the sum: {sum_norms[i]} ')

print("")

for i in range(len(norm_type)):
    print(f'norm {norm_names[i]}: {"is valid" if (sum_norms[i] <= u_norms[i] + v_norms[i]) else "is not valid"} for the sum of u and v')
```
