# Exercise 1




Create a 2 x 2 matrix and a 3 x 3 matrix.
Make sure they are both invertible - an easy way to do that is to make sure the determinant is not zero.

Test in python that the 3 properties listed on the slide with properties of the inverse holds for both these matrices.

```python
import numpy as np

np.random.seed(42)

while True:
    A2 = np.random.randint(10, size=(2,2))
    if np.linalg.det(A2) != 0:
        break


while True:
    A3 = np.random.randint(10, size=(3,3))
    if np.linalg.det(A3) != 0:
        break
    

print(f"A2 det: {np.linalg.det(A2)}")
print(f"A3 det: {np.linalg.det(A3)}\n")

print(f"A2 rank: {np.linalg.matrix_rank(A2)}")
print(f"A3 rank: {np.linalg.matrix_rank(A3)}\n")

print(f"A2 inverse: \n{np.linalg.inv(A2)}\n")
print(f"A3 inverse: \n{np.linalg.inv(A3)}\n")

print(f"A2*A2^-1 : \n{np.round(np.linalg.inv(A2)@A2, decimals=10)}\n")
print(f"A3*A3^-1 : \n{np.round(np.linalg.inv(A3)@A3, decimals=10)}\n")
```

Construct the inverse MANUALLY (i.e. not using np.linalg.inv library) and check that it is indeed an inverse.

What happens if you replace the number 3 with 0?
Try it out in python using np.linalg.inv - what happens?

```python
import numpy as np

M = np.array([
    [1,0,0],
    [0,2,0],
    [0,0,3]
])

invM = np.zeros((3,3))

for i in range(len(M)):
    if M[i,i] != 0:
        invM[i,i] = 1 / M[i,i]

print(invM)



M = np.array([
    [1,0,0],
    [0,2,0],
    [0,0,0]
])

invM = np.zeros((3,3))

for i in range(len(M)):
    if M[i,i] != 0:
        invM[i,i] = 1 / M[i,i]

print(invM)
print("it works cause i am good :D")
```

# Exercise 6

The following matrix is not a square matrix, so it does not have a full inverse. However, it does have a left inverse.
Construct the left inverse and test it is indeed a left inverse.

```python
import numpy as np

T = np.array([ [1,0],[2,1],[3,2]])
print(T)
print("")

print("T.T @ T: \n", T.T @ T)

left_inverse = np.linalg.inv(T.T@T)@T.T
print(left_inverse)
print("")
print(f"left inverse * T: \n{np.round(left_inverse@T,8)}\n")
```

# Exercise 7 (optional)

Derive the right-invefrse (not left!) from wide matrices by following the same method that derives the left inverse (and adjusting along as neccesary)

```python
import numpy as np

T = np.array([ [1,0],[2,1],[3,2]]).T
print(T)
print("")

print("T.T @ T: \n", T.T @ T)

right_inverse = T.T@np.linalg.inv(T@T.T)
print(right_inverse)
print("")
print(f"T * right inverse: \n{np.round(T@right_inverse,10)}\n")
```
