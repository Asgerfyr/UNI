give I (we are not using R matrix here):

 inverse Q @ Q, \

(the formular on page 155 in the book)

```python
Q,R  = np.linalg.qr( np.random.randn(5,5) )
"""
 Q.T @ Q, \
 Q @ Q.T, \
 inverse Q @ Q, \
  Q @ inverse Q, \
"""
print(np.round(Q.T @ Q, 5),"\n")
print(np.round(Q @ Q.T, 5),"\n")
print(np.round(np.linalg.inv(Q) @ Q, 5),"\n")
print(np.round(Q @ np.linalg.inv(Q), 5),"\n")
```

Indeed, check that is the case.

It might be difficult to visually inspect a 6x6 matrix, so an easy check is to calculate the norm of the difference of the two inverses - this gives a single number, which should be zero (or very close to zero). Try this in at least one of the 3 cases.


If one of the subtractions between the two inverses are not zero, try to come up with some suggestions for why they are not 100% identical.

```python
"""
Task
1. Create matrices
Generate 3 random matrices:
4×4
5×5
6×6
Ensure each matrix is invertible
Check determinant ≠ 0
If not, generate a new matrix
"""

matrixes = [0,0,0]
while True:
    matrixes[0] = np.random.rand(4,4)
    if np.linalg.det(matrixes[0]) != 0:
        break

while True:
    matrixes[1] = np.random.rand(5,5)
    if np.linalg.det(matrixes[1]) != 0:
        break

while True:
    matrixes[2] = np.random.rand(6,6)
    if np.linalg.det(matrixes[2]) != 0:
        break


"""
2. Compute inverses (2 methods)
For each matrix:
Find inverse using built-in Python method
Find inverse using QR decomposition
"""
inverse_builtin = [np.linalg.inv(matrix) for matrix in matrixes]

inverse_qr = [0,0,0]
for i, matrix in enumerate(matrixes):
    Q,R = np.linalg.qr(matrix)
    inverse_qr[i] = np.linalg.inv(R) @ Q.T

"""
3. Compare results
For each matrix:
Subtract the two inverses
Result should be ≈ zero matrix
"""
diffrence = [0,0,0]
for i in range(3):
    diffrence[i] = inverse_builtin[i] - inverse_qr[i]
    print(f"diffrence for matrix {i}:\n {np.round(diffrence[i], 5)}\n")
"""
4. Verify accuracy
Check that differences are close to zero
For at least one matrix:
Compute the norm of the difference
Should be ≈ 0
"""
norms = [0,0,0]
for i in range(3):
    norms[i] = np.linalg.norm(diffrence[i],ord="fro")
    print(f"diffrence for matrix {i}:\n {norms[i]}\n")
"""
5. Reflect
If results are not exactly zero:
Suggest reasons why (e.g. numerical errors)
"""
print("rounding errors") 
```
