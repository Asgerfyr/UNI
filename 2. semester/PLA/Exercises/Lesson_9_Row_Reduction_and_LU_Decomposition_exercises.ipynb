# Exercise 1

Solve the following equations using RREF and the augmented matrix technique using python.

5x + 4 = 2y

7y + 3 = 3x

After solving for x and y, then check if your solutions are correct using the math.isclose method to compare the right and the left side of each of the equations.

```python
"""
5x + 4 = 2y
7y + 3 = 3x

formattet in a uniform format:
5x - 2y = -4
-3x + 7y = -3
"""

M = np.array([[5, -2, -4],[-3, 7, -3]])

symMat = sym.Matrix(M)

rrefMat, pivotCols = symMat.rref()
print(f"RREF Matrix:\n{rrefMat}")
print(f"Pivot Columns: {pivotCols}")
x = rrefMat[0, 2]
y = rrefMat[1, 2]

print(f"Solution: x = {x}, y = {y}")

print("\nChecking solution:")
eq1_lhs = 5*x + 4
eq1_rhs = 2*y
eq2_lhs = 7*y + 3
eq2_rhs = 3*x
print(f"Equation 1: LHS = {eq1_lhs}, RHS = {eq1_rhs}")
print(f"Equation 2: LHS = {eq2_lhs}, RHS = {eq2_rhs}")

```

```python
"""
y = 3x + 5

2y - 20 =  6x

formattet in a uniform format:
-3x + 1y = 5
6x - 2y = -20
"""

M = np.array([[-3, 1, 5],[6, -2, -20]])

symMat = sym.Matrix(M)

rrefMat, pivotCols = symMat.rref()
print(f"RREF Matrix:\n{rrefMat}")
print(f"Pivot Columns: {pivotCols}")

if(np.linalg.matrix_rank(M) > np.linalg.matrix_rank(M[:, :-1])):
    print("The system has no solution.")
elif(np.linalg.matrix_rank(M) == np.linalg.matrix_rank(M[:, :-1]) == M.shape[1] - 1):
    print("The system has infinitely many solutions.")
    print(f"{rrefMat[0, 0]}x+{rrefMat[0, 1]}y={rrefMat[0, 2]}, {rrefMat[1, 0]}x+{rrefMat[1, 1]}y={rrefMat[1, 2]} + t")
else:
    print("The system has a unique solution.")


    x = rrefMat[0, 2]
    y = rrefMat[1, 2]

    print(f"Solution: x = {x}, y = {y}")

    print("\nChecking solution:")
    eq1_lhs = -3*x + 1*y
    eq1_rhs = 5
    eq2_lhs = 6*x - 2*y
    eq2_rhs = -20
    print(f"Equation 1: LHS = {eq1_lhs}, RHS = {eq1_rhs}")
    print(f"Equation 2: LHS = {eq2_lhs}, RHS = {eq2_rhs}")

```

One application of LU decomposition is computing the determinant.

There are two nice properties of the determinant (presented here without proof):

1. det(AB) = det(A)det(B)
2. The determinant of an upper or lower triangular matrix is the product of the numbers on the diagonal.

(so det(I) is in fact always 1 - why? )

So if A = LU, then det(A) = det(L)*det(U), and since
both L and U are triangular matrices we can compute det(A) as the product of all diagonal elements of L and U.

To get the diagonal elements you can use this in python:

np.diag(U), which returns an array of the diagonal.

To calculate the product of this array you can use this:
