```python
res_task_()
"""
--equations--
The sum of the flow rates through pipe x plus twice the flow rate through pipe y must equal 5 units per second minus twice the flow rate through pipe z
x+2y = 5 - 2z
The flow rate through pipe z must equal twice the flow rate through pipe x minus the flow rate through pipe y.
z = 2x - y
The sum of five times the flow rate through pipe x plus the flow rate througth pipe y minus the flow rate through pipe z must be equal to 3 units per second.
5x + y - z = 3
"""
#q1
task_(f"""
The equation system in a uniform format:
  x + 2y + 2z = 5
-2x +  y +  z = 0
 5x +  y + -z = 3
""")
print("The system of equations in a matrix:")

#coefficient
A = np.array([
    [1,2,2],
    [-2,1,1],
    [5,1,-1]
])

#constants
b = np.array([5,0,3])

M = np.hstack((A, b.reshape(-1,1)))

symMat = sym.Matrix(M)

print("augmented matrix:\n", symMat)

#q2
task_("The solution of the system of equations is:")
print("RREF from python:")
S = symMat.rref()[0]
print(S)
print("solution:\n", np.array([['x','y','z'],[*S[:,3]]]).T,"\n")



#q3
task_("""
in the RREF solution the coefficient part of the matrix is the I3 matrix witch means that all variables have a unique solution given by the constants part of the matrix
""")


#q4
task_()
rank_A = np.linalg.matrix_rank(A)
rank_aug = np.linalg.matrix_rank(M)  # M is the augmented matrix [A|b]
n_unknowns = A.shape[1]

print(f"rank A: = {rank_A}")
print(f"rank [A|b] = {rank_aug}")
print(f"number of unknowns = {n_unknowns}")


if rank_aug > rank_A:
  print(
    "0 solutions: rank([A|b]) > rank(A), ",
    "so b cannot be expressed as a linear combination of A's columns"
  )
elif rank_aug == rank_A == n_unknowns:
  print(
    "1 solution: rank([A|b]) = rank(A) = number of unknowns, ",
    "so all variables are a unique linear combination of A's columns"
  )
elif rank_aug == rank_A < n_unknowns:
  print(
    "Infinitely many solutions: rank([A|b]) = rank(A) < number of unknowns, "
    "so at least one variable is free. fx. x + y = 10"
  )

#q5
task_(f"""Since rank = equals the number of unknowns, the system has exactly one unique solution, witch match with the result from question 3.""")

#q6
task_()
result = S[:,3]
x, y, z = result
print(f"checking solution:\n")
print(f"equation 1: x + 2y = 5 - 2z")
print(f"right side: {x + 2*y} = left side: {5 - 2*z}")
print(f"equation 2: z = 2x - y")
print(f"right side: {z} = left side: {2*x - y}")
print(f"equation 3: 5x + y - z = 3")
print(f"right side: {5*x + y - z} = left side: {3}")
```

Outdoor temperature

Duration

End temperature

Methodology / Progression

This section describes how the regression model is constructed and why each step is necessary.

Continuous temperature data from November to present
Real-world conditions (no control over variables)

```python


from datetime import datetime

# get datapoints
data = pd.read_csv('fyr_temp.csv')
df = pd.DataFrame(data) #the data has temperature at index 0 and time at index 1
# keep only valid numeric temperature rows (filters out 'unavailable')
df['state'] = pd.to_numeric(df['state'], errors='coerce')
df = df.dropna(subset=['state', 'last_changed']).copy()

#make plot grids to visualize the data and regression results
fig, ax = plt.subplots(3, 2)

# make data into numpy array to work with it
RAW_DATA = df.drop(['entity_id'], axis=1).to_numpy()
ax[0, 0].scatter(RAW_DATA[:, 1], RAW_DATA[:, 0], s=1) # plot raw data to visualize
ax[0, 0].set_title("Raw data")
ax[0, 0].xaxis.set_visible(False) # removing the x axis values since they are not in a usable format
ax[0, 0].set_ylabel("temperature")

# filter out heating cycles
temp = RAW_DATA[:, 0]
filter_mask = temp[:-1] > temp[1:]
RAW_DATA = RAW_DATA[:-1][filter_mask]


# format the time into usable format
RAW_DATA[:, 1] = [datetime.fromisoformat(t.replace('Z', '+00:00')) for t in RAW_DATA[:, 1]]
start_time = RAW_DATA[0, 1].timestamp()
RAW_DATA[:, 1] = [t.timestamp() - start_time for t in RAW_DATA[:, 1]]#zero the time so the first datapoint is at t=0

# ensure numeric dtype before splitting/fitting
RAW_DATA = np.array(RAW_DATA, dtype=float)

# group the cooling cycle into segments
temp = RAW_DATA[:, 0]
split_idx = np.where(temp[1:] > temp[:-1])[0] + 1 #array of splitting indexes where the temperature from the last datapoint is lower then the current datapoint
segments = np.split(RAW_DATA, split_idx) #array of all cooling segments


#changing the time of each segment to start at t=0
unfittet_DATA = np.empty((0, 2), dtype=float)

for segment in segments:
    t0 = segment[0, 1]
    segment[:, 1] = segment[:, 1] - t0
    unfittet_DATA = np.concatenate((unfittet_DATA, segment))
    

# sort ing the data dy time to get a data set that more closely represent ond continuous cooling cycles
# this also makes it bettter for plotting grafs with the x axis of the data as time
unfittet_DATA = unfittet_DATA[np.argsort(unfittet_DATA[:, 1])]
ax[0, 1].scatter(unfittet_DATA[:, 1], unfittet_DATA[:, 0], s=1)
ax[0, 1].set_title("Unfittet data")
ax[0, 1].set_xlabel("time")
ax[0, 1].set_ylabel("temperature")


# since the data is from cooling cycles, we model it as k*e^(s*t)
# rewrite this as ln(y) = ln(k) + s*t lets us fit with linear regression as it is no longer an exponential model but a linear model in the log temperature space
# ln(y) = ln(k) + s*t lets us fit with linear regression
log_unfittet_DATA = unfittet_DATA.copy()
log_unfittet_DATA[:, 0] = np.log(log_unfittet_DATA[:, 0]) #taking the log of the temperature data

# plot the log temperature data to visualize the linear relationship
ax[1, 0].scatter(log_unfittet_DATA[:, 1], log_unfittet_DATA[:, 0], s=1)
ax[1, 0].set_title("Log of unfittet data")
ax[1, 0].set_xlabel("time")
ax[1, 0].set_ylabel("log(temperature)")

# linear regression on the log temperature data
X = np.array(log_unfittet_DATA[:, 1], dtype=float)
Y = np.array(log_unfittet_DATA[:, 0], dtype=float)
D_F = np.column_stack((np.ones(len(X)), X)) # design matrix for linear regression with an intercept term
Beta_hat_F = np.linalg.inv((D_F.T @ D_F)) @ D_F.T @ Y # calculating the regression coefficients


# As the coefficient for k is just the interception with t0, the only thing we are interested in is the slope
print(f"slope: s = {Beta_hat_F[1]}\n") # printing the slope of the regression model
k1 = np.exp(Beta_hat_F[0])
s1 = Beta_hat_F[1]


# defining the model function of the regression model found
def regression_model_1(t, k=k1):
    return k * np.exp(s1 * np.asarray(t, dtype=float))

# plot the graf found by the regression model
# plot the graf for the log temperature data
ax[1, 0].plot(X, D_F @ Beta_hat_F, color='red')

# plot the graf for the unfittet cooling cycle data
ax[0, 1].plot(X, regression_model_1(X), color='red')


#now we will use this regression model to fit the data better by moving the data points to the best place on the model
#aka manipulation the data points to fit the model better
# hopefully this will give us a tendency that better represents a real cooling cycle and not just a mix of different cooling cycles that are not aligned in time

# making a function to calculate the score group on a model
def score_model(group, function):
    group = np.array(group, dtype=float, copy=False) # making sure the group is a numpy array with the right dtype to do calculations on it
    return np.sum(group[:, 0] - function(group[:, 1]))/len(group) # calculating the score as the average distance from the data point to the model provided


# making a function to move the segment to the best place on the model
def move_segment(group, function, learning_rate=30, max_iterations=1000, tolerance=1e-13):
    for _ in range(max_iterations): # the cooling cycle get max_iterations chances to move to the best place on the model
        scores = score_model(group, lambda t: function(t)) # calculating the score for the current position of the cooling cycle
        group[:, 1] -= learning_rate * scores # moving the time of the cooling cycle by a factor of the score to try to get it closer to the model
        if np.abs(scores) < tolerance: # if the score is close enough to 0 we assume we have found the best place and further iterations are not needed
            break
    return group


# making an empty array to store the fittet data in
fittet_DATA = np.empty((0, 2), dtype=float)

# moving each segment to the best place on the model and storing the fittet data in fittet_DATA
for segment in segments:
    fittet_DATA = np.concatenate((fittet_DATA, move_segment(segment, regression_model_1)))

# plot the fittet data to visualize the result of moving the data points to fit the model better
ax[1, 1].scatter(fittet_DATA[:, 1], fittet_DATA[:, 0], s=1)
ax[1, 1].set_title("Fittet data")
ax[1, 1].set_xlabel("time")
ax[1, 1].set_ylabel("temperature")


# sort
fittet_DATA = fittet_DATA[np.argsort(fittet_DATA[:, 1])]
ax[1, 1].plot(fittet_DATA[:, 1], regression_model_1(fittet_DATA[:, 1]), color='red')


#just as the previous regression model was a linear regression on the log temperature data, we can do the same for the fittet data
#and this should hopefully give a closer slope to the real cooling cycle
log_fittet_DATA = fittet_DATA.copy()
log_fittet_DATA[:, 0] = np.log(log_fittet_DATA[:, 0])

# plot the log of the fittet data to visualize the linear relationship
ax[2, 0].scatter(log_fittet_DATA[:, 1], log_fittet_DATA[:, 0])
ax[2, 0].set_title("Log of fittet data")
ax[2, 0].set_xlabel("time")
ax[2, 0].set_ylabel("log(temperature)")

# linear regression on the new log-transformed data
X = np.array(log_fittet_DATA[:, 1], dtype=float)
Y = np.array(log_fittet_DATA[:, 0], dtype=float)
D_F = np.column_stack((np.ones(len(X)), X))
Beta_hat_F = np.linalg.inv((D_F.T @ D_F)) @ D_F.T @ Y


print(f"slope: s = {Beta_hat_F[1]}\n") # printing the slope of the regression model

# as the coefficient for k is just the interception with t0, the only thing we are interested in is the slope
s2 = Beta_hat_F[1]
k2 = np.exp(Beta_hat_F[0])


# defining the model function of the new regression model found
def regression_model_2(t, k=k2):
    return k * np.exp(s2 * np.asarray(t, dtype=float))


# plot the graf found by the regression model
ax[1, 1].plot(fittet_DATA[:, 1], regression_model_2(fittet_DATA[:, 1]), color='black', linestyle='--')


# plot log-fit and transformed fit
ax[2, 0].plot(X, D_F @ Beta_hat_F, color='black')



#getting the data of the new
TC_data = pd.read_csv('todays_cooling.csv') 
tc_df = pd.DataFrame(TC_data) #the data has temperature at index 0 and time at index 1

#format the Data and time into usable format
tc_RAW_DATA = tc_df.drop(['entity_id'], axis=1).to_numpy()
tc_RAW_DATA[:, 1] = [datetime.fromisoformat(t.replace('Z', '+00:00')) for t in tc_RAW_DATA[:, 1]]
start_time = tc_RAW_DATA[0, 1].timestamp()
tc_RAW_DATA[:, 1] = [t.timestamp() - start_time for t in tc_RAW_DATA[:, 1]]#zero the time so the first datapoint is at t=0

ax[2, 1].scatter(tc_RAW_DATA[:, 1], tc_RAW_DATA[:, 0], s=1)
ax[2, 1].set_title("Todays cooling cycle")
ax[2, 1].set_xlabel("time")
ax[2, 1].set_ylabel("temperature")


#the k value is the intersection with the y axis
#in our case it is the starting temperature of the cooling cycle, 
k = tc_RAW_DATA[0, 0]

x = range(0, 10000, 10)
ax[2, 1].plot(x, regression_model_1(x,k=k), color='red')
ax[2, 1].plot(x, regression_model_2(x,k=k), color='black', linestyle='--')

plt.tight_layout()
plt.show()
```

The regression model does not accurately represent the most recent cooling cycle. The deviation is significant, and several factors may contribute to this mismatch.

One possible factor is the use of straw instead of wood as fuel. However, the primary issue is likely the quality of the dataset. The data used for the regression was collected over a long period and compressed to one data point per hour to reduce storage requirements. This low temporal resolution distorts short-term processes.

To improve accuracy, a new regression will be based on uncompressed data with higher time resolution. This dataset spans approximately 10 days and allows for a more precise representation of the cooling cycle, resulting in a model that better matches current observations.
