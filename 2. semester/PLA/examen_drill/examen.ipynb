# Singular Value Decomposition: A = UΣV^T
# U: orthogonal matrix (left singular vectors)
# Σ: diagonal matrix of singular values
# V^T: orthogonal matrix (right singular vectors)
U, s, VT = np.linalg.svd(A)
# Verify U is orthogonal
assert np.allclose(U.T @ U, np.eye(U.shape[1]))
# Verify V is orthogonal
assert np.allclose(VT @ VT.T, np.eye(VT.shape[0]))
# Verify reconstruction
Σ = np.diag(s)
assert np.allclose(A, U @ Σ @ VT)
