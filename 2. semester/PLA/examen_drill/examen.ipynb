# Question 2

```python


#task 1
print(f"{"\n\n"}\033[1;32m#task 1\033[0m")
A = np.array([
    [2, 1, 0, 3],
    [0, 4, 1, 2],
    [1, 0, 3, 1],
    [3, 2, 1, 0],
    [1, 1, 2, 4]
])

print(f"A = \n{A}")

#task 2
print(f"{"\n\n"}\033[1;32m#task 2.1\033[0m")

# calculate A^T A
A_T = A.T

A_T_A = A_T@A

print(f"A^t A = {A_T_A}")

# Calculate trace: sum of diagonal elements
display(Math(r"\textbf{Trace:} \quad \text{tr}(\mathbf{A^t A}) = \sum_{i=1}^{n} a_{ii}"))

A_T_A_trace = np.trace(A_T_A)

print(f"trace(A^t A) = {A_T_A_trace}")


# Frobenius norm
#task 2.2
print(f"{"\n\n"}\033[1;32m#task 2.1\033[0m")

display(Math(r"\textbf{Matrix Rank:} \quad \text{rank}(\mathbf{A}) = \text{number of linearly independent rows/columns}"))

A_T_A_norm_fro = np.linalg.norm(A_T_A, ord='fro')

print(f"fro_norm(A^t A) = {A_T_A_norm_fro}")


#task 3
print(f"{"\n\n"}\033[1;32m#task 3\033[0m")
print(f"\n{'='*50}")
print("The number of linearly independent columns of A can be at most 4, since A has 4 columns (each vector lies in a 5-dimensional space).")
print("There are no obvious linearly dependent columns, which suggests full column rank.")
print(f"If all vectors were linearly dependent, this would lead to the norm and trace being equal, which they are not.")
print("This supports the conclusion that the rank is 4, since there is no evidence of linear dependence.")
print("However, to confirm this, a rank test is preformed.")
print(f"{'='*50}\n")
A_rank = np.linalg.matrix_rank(A)
print(f"A_rank = {A_rank}")  
```

```python
#task 1
print(f"{"\n\n"}\033[1;32m#task 1\033[0m")

display(Math(r"\textbf{Left Inverse:} \quad \mathbf{A}_{\text{left}}^{-1} = (\mathbf{A}^T \mathbf{A})^{-1} \mathbf{A}^T"))

R = np.linalg.inv(A_T_A) @ A_T

print(R)


#task 2
print(f"{"\n\n"}\033[1;32m#task 2\033[0m")

display(Math(r"\textbf{Property:} \quad \mathbf{A}_{\text{left}}^{-1} @ \mathbf{A} = \mathbf{I}"))

# Verify: should be identity matrix
print(np.round(R @ A, 8))

print("\n The left inverse exists so there is no need to make the right inverse")
```

# Question 4

```python
print(f"{"\n\n"}\033[1;32m#Question 4\033[0m")

display(Math(r"\textbf{System of Linear Equations:} \quad \mathbf{Ax} = \mathbf{b}"))

# 2x + y - z + w = 4
# x - y + 2z + w = 1
# 3x + 2y + z - w = 10
# x + y + z + w = 6

# 2x + 1y − 1z + 1w = 4
# 1x − 1y + 2z + 1w = 1
# 3x + 2y + 1z − 1w = 10
# 1x + 1y + 1z + 1w = 6


A = np.array([
    [2, 1, -1, 1],
    [1, -1, 2, 1],
    [3, 2, 1, -1],
    [1, 1, 1, 1]
])
b = np.array([4, 1, 10, 6])

display(Math(r"\textbf{Augmented Matrix:} \quad [\mathbf{A} | \mathbf{b}]"))

M = np.hstack((A, b.reshape(-1, 1)))

symMat = sym.Matrix(M)
rrefMat, pivotCols = symMat.rref()

print(f"RREF Matrix:\n{rrefMat}")
print(f"Pivot Columns: {pivotCols}")

(x,y,z,w) = rrefMat[:,4]


print(f"x = {x} = {float(x):.4f}")
print(f"y = {y} = {float(y):.4f}")
print(f"z = {z} = {float(z):.4f}")
print(f"w = {w} = {float(w):.4f}")
```

```python
#task 1
print(f"{"\n\n"}\033[1;32m#task 1\033[0m")

B = np.array([
    [3, 1, 0],
    [0, 3, 0],
    [0, 1, 5]
    ])

display(Math(r"\textbf{Diagonalization:} \quad \mathbf{A} = \mathbf{V} \mathbf{D} \mathbf{V}^{-1}"))


eigenvalues, eigenvectors = np.linalg.eig(B)

V = eigenvectors
D = np.diag(eigenvalues)
V_inv = np.linalg.inv(V)

print(f"V =\n{np.round(V, 2)}")
print(f"D =\n{np.round(D, 2)}")
print(f"V^-1 =\n{np.round(V_inv, 2)}")

print("\n")

print("Rank of eigenvector matrix:", np.linalg.matrix_rank(eigenvectors))
print("Matrix size:", B.shape[0])

print("The eigenvector matrix has rank 2, meaning it contains only 2 linearly independent eigenvectors.")
print("A 3x3 matrix requires 3 linearly independent eigenvectors to be diagonalizable.")
print("Since there are only 2 independent eigenvectors, the matrix is not diagonalizable.")

#task 2
print(f"{"\n\n"}\033[1;32m#task 2\033[0m")

display(Math(r"\textbf{Property:} \quad \mathbf{A}^n = \mathbf{V} \mathbf{D}^n \mathbf{V}^{-1}"))

# Verify: V @ D @ V^-1 should equal B
B_reconstructed = V @ D @ V_inv

print(f"B reconstructed =\n{np.round(B_reconstructed, 2)}")

print(f"Reconstruction correct: {np.allclose(B, B_reconstructed)}")

print(f"The matrix chould not be recunstructed")
```

```python
#task 1
print(f"{"\n\n"}\033[1;32m#task 1\033[0m")

display(Math(r"$\mathbf{A} = \mathbf{U} \mathbf{\Sigma} \mathbf{V}^T$"))

U, s, Vt = np.linalg.svd(B)

# Construct Sigma matrix from singular values
Sigma = np.zeros_like(B, dtype=float)
np.fill_diagonal(Sigma, s)

print(f"U:\n{np.round(U, 2)}\n")
print(f"S (singular values):\n{np.round(s, 2)}\n")
print(f"V^T:\n{np.round(Vt, 2)}")



#task 2
print(f"{"\n\n"}\033[1;32m#task 2\033[0m")

# Verify: U @ Sigma @ Vt should equal B
B_reconstructed = U @ Sigma @ Vt

print(f"B reconstructed =\n{np.round(B_reconstructed, 2)}")
print(f"Reconstruction correct: {np.allclose(B, B_reconstructed)}")
```

# Question 7

```python
X = np.array([
    [3, 8, 68],
    [5, 10, 78],
    [7, 6, 82],
    [6, 12, 90],
    [8, 9, 88],
    [4, 5, 65],
    [9, 11, 95],
    [5, 7, 74]
])


#task 1
print(f"{"\n\n"}\033[1;32m#task 1\033[0m")

# Apply data to named var
hours = X[:,0]
completed = X[:,1]
scores = X[:,2]


# Design matrix: add intercept column
X_design = np.column_stack((np.ones(hours.shape[0]), hours, completed))
y = scores.astype(float)

# Solve least squares: X_design @ beta = y
beta, residuals, rank, svals = np.linalg.lstsq(X_design, y, rcond=None)

beta0, beta1, beta2 = beta
print(f"Model: y = beta0 + beta1*x1 + beta2*x2")
print(f"Model: y = {beta0:.4f} + {beta1:.4f}*x1 + {beta2:.4f}*x2")
print(f"Model: y = {beta0:.4f} + {beta1:.4f}*hours + {beta2:.4f}*completed")



#task 2
print(f"{"\n\n"}\033[1;32m#task 2\033[0m")

# Predict for a new sample
hours_pred = 6
completed_pred = 8.0
y_pred = beta0 + beta1 * hours_pred + beta2 * completed_pred
print(f"Predicted score for hours={hours_pred}, completed={completed_pred}: {y_pred:.4f}")


#task 3
print(f"{"\n\n"}\033[1;32m#task 3\033[0m")
# SSE (Sum of Squared Errors)
display(Math(r"\textbf{SSE:} \quad \mathrm{SSE} = \sum_{i=1}^{m}(y_i - \hat{y}_i)^2"))
y_fitted = X_design @ beta
SSE = np.sum((y - y_fitted) ** 2)
print(f"SSE: {SSE:.4f}")
```

b is the model that we want to explain the data with and this model doesn't land inside the column space

to fix this problem b is projected onto the column space to find the best possible approximation of the model

this is the core idea in least squares where we want to minimize for error by getting a model that is as close to the column space of A as possible, and this is the projection of b onto the column space of A
