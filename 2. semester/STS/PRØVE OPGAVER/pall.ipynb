For this, the formula for total probability is used.
```python
# Total sandsynlighed med to komplementære hændelser
# P(B) = P(A) × P(B|A) + P(A^c) × P(B|A^c)
# P(B) = P(A) × P(B|A) + (1-P(A)) × P(B|A^c)

P_B = P_B_T  # P(A) - sandsynlighed for hændelse A
P_P_B = P_P_B  # P(B|A)
P_P_nB = P_P_nB  # P(B|A^c)

# Total sandsynlighed
P_P = P_B * P_P_B + (1 - P_B) * P_P_nB
print(f"P(P) = {P_P:.4f}")
print(round(P_P*100, 2), "%")
```

En vindmølle producent ønsker at undersøge om deres nye vindmølle design genererer mere
eller mindre elektrisk energi per år end deres gamle vindmølle. De sætter derfor en ny

kan antages, at vindforhold og lignende er ens for hver ny og gammel vindmølle.
Vi kan antage, at forskellene i genereret energi per år mellem ny og gammel vindmølle på
samme lokation følger en normalfordeling. Nedenstående tabel Viser energimængderne per år
for de 10 nye og 10 gamle vindmøller:


Vi ønsker at undersøge om de nye og gamle vindmøller producerer samme mængde energi

### A Opstil et udtryk for normalfordelingen for effektforskellen mellem de nye og gamle vindmøller

```python
import numpy as np
from scipy import stats
from scipy.stats import norm, t, chi2, f, binom, poisson, expon, uniform, bernoulli, geom
import matplotlib.pyplot as plt

gammel = np.array([11.3,17.4,12.4,14.5,23.5,19.6,10.5,40.7,35.6,42.5])
ny = np.array([12,19.7,14.8,15.4,28.2,22.2,13.5,42.7,38.6,45.8])

diff = ny-gammel

mean = np.mean(diff)

# Variance from simulated data
variance = np.sum((diff - np.mean(diff)) ** 2) / (len(diff) - 1)
print(f"Variance: {variance:.6f}")


mu = mean
sigma2 = variance

print(f"D = N(mu_diff, sigma^2_diff) = N({mu:.2f},{sigma2:.2f})")
```

### B Opskriv en nulhypotese og alternativ hypotese for testen

$ H_0: \mu_D =0 $ den nye og den gamle vindmølle generare samme mængde energi

$ H_1: \mu_D \neq 0 $ den nye og den gamle vindmølle generare ikke samme mængde energi

### C Opstil udtryk for og beregn p-værdien. Kan vi afvise nulhypotesen på 5% signifikansniveau?

```python
import numpy as np
from scipy import stats
from scipy.stats import norm, t, chi2, f, binom, poisson, expon, uniform, bernoulli, geom
import matplotlib.pyplot as plt

# t-test for middelværdi (ukendt varians)
# H₀: μ = μ₀  vs  H₁: μ ≠ μ₀
sample_mean = mean
mu_0 = 0
s = np.sqrt(variance)  # Stikprøvestandardafvigelse
n = 10

# Teststatistik: t = (x̄ - μ₀) / (s/√n)
t_stat = (sample_mean - mu_0) / (s / np.sqrt(n))
df = n - 1

alpha = 0.05
t_critical = stats.t.ppf(1 - alpha/2, df)
p_value = 2 * (1 - stats.t.cdf(abs(t_stat), df))

print(f"H₀: μ = {mu_0}")
print(f"H₁: μ ≠ {mu_0}")
print(f"t({df}) = {t_stat:.4f}")
print(f"Kritisk værdi: ±{t_critical:.4f}")
print(f"p-værdi = {p_value:.7f}")
print(f"Konklusion: {'Forkast H₀' if abs(t_stat) > t_critical else 'Forkast ikke H₀'}")
```

### D Opstil udtryk for og beregn 99% konfidensintervallet for forskellen i genereret energi mellem nye og gamle vindmøller.

```python
# Konfidensinterval for middelværdi μ (ukendt populationsvarians)
# BRUG NÅR: Populationsstandardafvigelse σ er UKENDT - bruger stikprøve-std s
# Bruger T-FORDELING med df = n-1 frihedsgrader
# Antager: Data er tilnærmelsesvis normalfordelt
# CI: x̄ ± t_{α/2,n-1} × (s/√n)
# LaTeX: CI = \left[\bar{x} - t_{\alpha/2,n-1}\frac{s}{\sqrt{n}}, \bar{x} + t_{\alpha/2,n-1}\frac{s}{\sqrt{n}}\right]
alpha = 0.01  # Signifikansniveau
sample_mean = mean  # x̄ - Observeret stikprøvemiddelværdi
s = s  # s - Stikprøvestandardafvigelse (beregnet fra data)
n = 10  # Stikprøvestørrelse

df = n - 1  # Frihedsgrader
t_critical = stats.t.ppf(1 - alpha/2, df)
SE = s / np.sqrt(n)
margin_error = t_critical * SE

lower_bound = sample_mean - margin_error
upper_bound = sample_mean + margin_error

print(f"{(1-alpha)*100:.0f}% Konfidensinterval: ({lower_bound:.4f}, {upper_bound:.4f})")
print(f"Frihedsgrader: {df}")
```

### E Fortolk resulatet i opgave D og vurder om denne understøtter konklusionen fra opgave 

Ud fra resultatet i opgave D kan der observeres, at nulhypotesen ligger uden for dette konfidensinterval, hvilket yderligere verificerer forkastelsen af nulhypotesen.

The linear regression model builds upon finding the minimizing the error between a straight line and the data points given. The formula can be seen underneath. And this is, there's a formula for finding the best approximate slope and the best approximate slope and the best possible axis intercept.
$$ \hat{\beta}_1 = \frac{\sum(x_i-\bar{x})(y_i-\bar{y})}{\sum(x_i-\bar{x})^2}, \quad \hat{\beta}_0 = \bar{y} - \hat{\beta}_1\bar{x} $$

This here uses some common components, this being SXX and SXY, and these are, for example, SXX is the sum of the X difference squared, and the difference being the difference between the mean and the actual value. So all these differences squared and summed up.

### B Beregn modellens parametre (hældningen af den lineære model og skæringen med y-aksen). Angiv hvorledes disse er beregnet

These are calculated from the formula presented in the previous exercise or the previous task, Task 2a.

```python
# Lineær regression - manuel beregning med formler
# Model: Y = β₀ + β₁X + ε where ε ~ N(0,σ²)
# LaTeX: \hat{\beta}_1 = \frac{S_{XY}}{S_{XX}} = \frac{\sum(x_i-\bar{x})(y_i-\bar{y})}{\sum(x_i-\bar{x})^2}
# LaTeX: \hat{\beta}_0 = \bar{y} - \hat{\beta}_1\bar{x}
# LaTeX: R^2 = \frac{S_{XY}^2}{S_{XX} \cdot S_{YY}} = 1 - \frac{\sum(y_i-\hat{y}_i)^2}{\sum(y_i-\bar{y})^2}
x = Vægt
y = Spænding

# Beregn middelværdier
x_bar = np.mean(x)
y_bar = np.mean(y)

# Beregn sum of squares og cross-products
# S_XX = Σ(xᵢ-x̄)²
S_XX = np.sum((x - x_bar) ** 2)

# S_YY = Σ(yᵢ-ȳ)²
S_YY = np.sum((y - y_bar) ** 2)

# S_XY = Σ(xᵢ-x̄)(yᵢ-ȳ)
S_XY = np.sum((x - x_bar) * (y - y_bar))

# Beregn regressionskoefficienter
# β̂₁ = S_XY / S_XX (hældning/slope)
beta_1 = S_XY / S_XX

# β̂₀ = ȳ - β̂₁·x̄ (skæring/intercept)
beta_0 = y_bar - beta_1 * x_bar

# Fitted values: ŷ = β̂₀ + β̂₁x
y_hat = beta_0 + beta_1 * x

# Beregn R² (forklaringsgrad)
# Metode 1: R² = S_XY² / (S_XX · S_YY)
R_squared = (S_XY ** 2) / (S_XX * S_YY)

# Metode 2: R² = 1 - SS_res/SS_tot
SS_res = np.sum((y - y_hat) ** 2)  # Residual sum of squares
SS_tot = S_YY  # Total sum of squares
R_squared_alt = 1 - (SS_res / SS_tot)

# Regressionsligningen
print(f'Regressionsmodel: ŷ = {beta_0:.4f} + {beta_1:.4f}x')
print(f'β̂₀ (intercept) = {beta_0:.4f}')
print(f'β̂₁ (slope) = {beta_1:.4f}')
print(f'R² = {R_squared:.4f}')
print(f'')
print(f'Sum of squares:')
print(f'S_XX = {S_XX:.4f}')
print(f'S_YY = {S_YY:.4f}')
print(f'S_XY = {S_XY:.4f}')
```

### E Brug hypotesetesten

$$ H_0: \beta_1 = 0 $$
$$ H_1: \beta_1 \neq 0 $$

```python
# T-test for regression slope: Test om der er lineær sammenhæng
# H₀: β₁ = 0 (ingen lineær sammenhæng)
# H₁: β₁ ≠ 0 (der er lineær sammenhæng)
# LaTeX: t = \frac{\hat{\beta}_1 - \beta_{1,0}}{SE(\hat{\beta}_1)} = \frac{\hat{\beta}_1}{\sqrt{s_r^2 / S_{XX}}}
# LaTeX: s_r^2 = \frac{\sum(y_i - \hat{y}_i)^2}{n-2} = \frac{SS_{res}}{n-2}
# LaTeX: SE(\hat{\beta}_1) = \sqrt{\frac{s_r^2}{S_{XX}}}, \quad df = n-2
# LaTeX: p = 2P(T_{n-2} > |t|) \text{ (tosidet)}


# Beregn middelværdier og S_XX
x_bar = np.mean(x)
y_bar = np.mean(y)
n = len(x)

# S_XX = Σ(xᵢ-x̄)²
S_XX = np.sum((x - x_bar) ** 2)

# S_XY = Σ(xᵢ-x̄)(yᵢ-ȳ)
S_XY = np.sum((x - x_bar) * (y - y_bar))

# Beregn β̂₁ (hældning)
beta_1_hat = S_XY / S_XX

# Beregn β̂₀ (skæring)
beta_0_hat = y_bar - beta_1_hat * x_bar

# Fitted values: ŷ = β̂₀ + β̂₁x
y_hat = beta_0_hat + beta_1_hat * x

# Residualer: eᵢ = yᵢ - ŷᵢ
residuals = y - y_hat

# Residual sum of squares: SS_res = Σ(yᵢ - ŷᵢ)²
SS_res = np.sum(residuals ** 2)

# Residual variance: s_r² = SS_res / (n-2)
df = n - 2  # Frihedsgrader for regression
s_r_squared = SS_res / df
s_r = np.sqrt(s_r_squared)

# Standard error af β̂₁: SE(β̂₁) = √(s_r²/S_XX)
SE_beta_1 = np.sqrt(s_r_squared / S_XX)

# Teststatistik: t = (β̂₁ - 0) / SE(β̂₁)
beta_1_0 = 0  # Nulhypotese: β₁ = 0
t_stat = (beta_1_hat - beta_1_0) / SE_beta_1

# p-værdi (tosidet test)
p_value = 2 * (1 - stats.t.cdf(abs(t_stat), df))

# Kritisk værdi
alpha = 0.05
t_critical = stats.t.ppf(1 - alpha/2, df)

print('=== T-test for lineær sammenhæng ===')
print(f'H₀: β₁ = {beta_1_0}')
print(f'H₁: β₁ ≠ {beta_1_0}')
print(f'')
print(f'β̂₁ (hældning) = {beta_1_hat:.6f}')
print(f'SE(β̂₁) = {SE_beta_1:.6f}')
print(f's_r² (residual varians) = {s_r_squared:.6f}')
print(f'S_XX = {S_XX:.4f}')
print(f'')
print(f't({df}) = {t_stat:.4f}')
print(f'Kritisk værdi (α={alpha}): ±{t_critical:.4f}')
print(f'p-værdi = {p_value}')
print(f'')
conclusion = 'Forkast H₀' if abs(t_stat) > t_critical else 'Forkast ikke H₀'
print(f'Konklusion: {conclusion}')
if abs(t_stat) > t_critical:
    print(f'Der er signifikant lineær sammenhæng mellem X og Y (p={p_value:.6f})')
else:
    print(f'Ingen signifikant lineær sammenhæng (p={p_value:.6f})')
```

### E Antag at p=0.18 og benyt CLT til at beregne sandsynligheden for at finde mere end 54 fej

**Centralgrænseværdisætningen (CLT):**

Givet X_1, X_2, \ldots, X_n$ uafhængige med E[X_i] = \mu$ og $\text{Var}(X_i) = \sigma^2$

**Stikprøvemiddelværdi:**
$$ \bar{X} = \frac{1}{n}\sum_{i=1}^{n} X_i \xrightarrow{d} N\left(\mu, \frac{\sigma^2}{n}\right) \text{ når } n \to \infty $$

**Binomial approximation:**
$$ X \sim \text{Binomial}(n, p) \approx N(np, np(1-p)) $$
når np \geq 5$ og n(1-p) \geq 5$

**Proportion approximation:**
$$ \hat{p} = \frac{X}{n} \approx N\left(p, \frac{p(1-p)}{n}\right) $$

**Standardisering:**
$$ Z = \frac{\bar{X} - \mu}{\sigma/\sqrt{n}} \xrightarrow{d} N(0,1) $$

```python
import numpy as np
from scipy import stats
from scipy.stats import norm, t, chi2, f, binom, poisson, expon, uniform, bernoulli, geom
import matplotlib.pyplot as plt

# Central Limit Theorem: Binomial → Normal approximation
# X ~ Binomial(n, p) ≈ N(np, np(1-p)) når n er stor
# Regel: Brug når np ≥ 5 og n(1-p) ≥ 5

n = 3000  # Antal forsøg
p = 0.018  # Succesandsynlighed

# Check CLT betingelser
print(f"np = {n*p:.1f}, n(1-p) = {n*(1-p):.1f}")
print(f"CLT gyldig: {n*p >= 5 and n*(1-p) >= 5}")

# Binomial parametre → Normal parametre
mu = n * p
sigma = np.sqrt(n * p * (1 - p))

print(f"X ~ Binomial({n}, {p}) ≈ N({mu:.2f}, {sigma**2:.2f})")
print(f"μ = {mu:.2f}, σ = {sigma:.4f}")

# Beregn sandsynlighed: P(X > k) med kontinuitetskorrektion
k = 54  # Kritisk værdi

# Med kontinuitetskorrektion: P(X > k) = P(X ≥ k+1) ≈ P(Z > (k+0.5-μ)/σ)
z = (k + 0.5 - mu) / sigma
p_greater = 1 - stats.norm.cdf(z)

print(f"")
print(f"P(X > {k}) med kontinuitetskorrektion:")
print(f"z = ({k}+0.5-{mu:.2f})/{sigma:.4f} = {z:.4f}")
print(f"P(X > {k}) = P(Z > {z:.4f}) = {p_greater:.6f}")
```

### C): Opstil et udtryk for den poolede varians og beregn denne

$$ s_p^2 = \frac{(n_1-1)s_1^2+(n_2-1)s_2^2}{n_1+n_2-2} $$

$$ s_1^2 = \frac{1}{n_1-1}\sum_{i=1}^{n_1}(x_{1i}-\bar{x}_1)^2 $$
$$ s_2^2 = \frac{1}{n_2-1}\sum_{i=1}^{n_2}(x_{2i}-\bar{x}_2)^2 $$

$$ s_p^2 = \frac{\sum_{i=1}^{n_1}(x_{1i}-\bar{x}_1)^2+\sum_{i=1}^{n_2}(x_{2i}-\bar{x}_2)^2}{n_1+n_2-2} $$

```python

n1 = n_ta
s1_sq = np.sum((ta - np.mean(ta))**2)/(n1-1)


n2 = n_ny
s2_sq = np.sum((ny - np.mean(ny))**2)/(n2-1)



# Pooled variance for two samples
# LaTeX: s_p^2 = \frac{(n_1-1)s_1^2 + (n_2-1)s_2^2}{n_1+n_2-2}
s_pooled_sq = ((n1 - 1) * s1_sq + (n2 - 1) * s2_sq) / (n1 + n2 - 2)
s_pooled = np.sqrt(s_pooled_sq)
print(f"Pooled variance: {s_pooled_sq:.4f}")
print(f"Pooled std: {s_pooled:.4f}")
```

### D): Opstil udtryk for og beregn p-værdien. Kan vi afvise nulhypotesen på 5% signifikansnivea

```python
# P-værdi beregning for to-stikprøve t-test med pooled varians
# LaTeX: t = \frac{(\bar{x}_1 - \bar{x}_2) - (\mu_1 - \mu_2)}{s_p\sqrt{\frac{1}{n_1} + \frac{1}{n_2}}}
# LaTeX: p = 2P(T_{df} > |t|) = 2(1 - \text{CDF}_{t,df}(|t|))

# Data fra to grupper
x1 = ny
x2 = ta

# Beskrivende statistik
n1, n2 = len(x1), len(x2)
x1_bar = np.mean(x1)
x2_bar = np.mean(x2)


# Pooled varians: s_p^2
s_p_sq = s_pooled_sq
s_p = s_pooled

# Teststatistik (generelt): t = ((x̄₁ - x̄₂) - (μ₁ - μ₂)) / (s_p√(1/n₁ + 1/n₂))
mu_diff = 0  # (μ₁ - μ₂) under H₀, typisk 0
t_stat = ((x1_bar - x2_bar) - mu_diff) / (s_p * np.sqrt(1/n1 + 1/n2))

# Frihedsgrader
df = n1 + n2 - 2

# P-værdi (tosidet): p = 2P(T_df > |t|) = 2(1 - CDF_t(|t|))
p_value = 1 * (1 - stats.t.cdf(abs(t_stat), df))

print(f"x̄₁ = {x1_bar:.4f}, x̄₂ = {x2_bar:.4f}")
print(f"s_p² = {s_p_sq:.4f}, s_p = {s_p:.4f}")
print(f"t({df}) = {t_stat:.4f}")
print(f"p-værdi = {p_value}")
```

### B): Beregn modellens parametre (hældning og skæring med y-aksen), og angiv hvordan disse er beregne

$$ \hat{\beta}_1 = \frac{S_{XY}}{S_{XX}} = \frac{\sum(x_i-\bar{x})(y_i-\bar{y})}{\sum(x_i-\bar{x})^2} $$

```python
# Lineær regression - manuel beregning med formler
# Model: Y = β₀ + β₁X + ε where ε ~ N(0,σ²)
# LaTeX: \hat{\beta}_1 = \frac{S_{XY}}{S_{XX}} = \frac{\sum(x_i-\bar{x})(y_i-\bar{y})}{\sum(x_i-\bar{x})^2}
# LaTeX: \hat{\beta}_0 = \bar{y} - \hat{\beta}_1\bar{x}
# LaTeX: R^2 = \frac{S_{XY}^2}{S_{XX} \cdot S_{YY}} = 1 - \frac{\sum(y_i-\hat{y}_i)^2}{\sum(y_i-\bar{y})^2}
x = n_kald
y = ms_responce

# Beregn middelværdier
x_bar = np.mean(x)
y_bar = np.mean(y)

# Beregn sum of squares og cross-products
# S_XX = Σ(xᵢ-x̄)²
S_XX = np.sum((x - x_bar) ** 2)

# S_YY = Σ(yᵢ-ȳ)²
S_YY = np.sum((y - y_bar) ** 2)

# S_XY = Σ(xᵢ-x̄)(yᵢ-ȳ)
S_XY = np.sum((x - x_bar) * (y - y_bar))

# Beregn regressionskoefficienter
# β̂₁ = S_XY / S_XX (hældning/slope)
beta_1 = S_XY / S_XX

# β̂₀ = ȳ - β̂₁·x̄ (skæring/intercept)
beta_0 = y_bar - beta_1 * x_bar

# Fitted values: ŷ = β̂₀ + β̂₁x
y_hat = beta_0 + beta_1 * x

# Beregn R² (forklaringsgrad)
# Metode 1: R² = S_XY² / (S_XX · S_YY)
R_squared = (S_XY ** 2) / (S_XX * S_YY)

# Metode 2: R² = 1 - SS_res/SS_tot
SS_res = np.sum((y - y_hat) ** 2)  # Residual sum of squares
SS_tot = S_YY  # Total sum of squares
R_squared_alt = 1 - (SS_res / SS_tot)

# Regressionsligningen
print(f'Regressionsmodel: ŷ = {beta_0:.4f} + {beta_1:.4f}x')
print(f'β̂₀ (intercept) = {beta_0:.4f}')
print(f'β̂₁ (slope) = {beta_1:.4f}')
print(f'R² = {R_squared:.4f}')
print(f'')
print(f'Sum of squares:')
print(f'S_XX = {S_XX:.4f}')
print(f'S_YY = {S_YY:.4f}')
print(f'S_XY = {S_XY:.4f}')
```

### F): Udfør en hypotesetest for signifikansen af hældningskoefficiente

𝐻0: 𝛽1 = 0 og 𝐻1: 𝛽1 ≠ 0

Vurder, om der er statistisk evidens for en lineær sammenhæng mellem belastning og responstid ved et signifikansniveau på 5%.

**Hældning:**
$$ \hat{\beta}_1 = \frac{S_{XY}}{S_{XX}} = \frac{\sum(x_i-\bar{x})(y_i-\bar{y})}{\sum(x_i-\bar{x})^2} $$

**Fitted values:**
$$ \hat{y}_i = \hat{\beta}_0 + \hat{\beta}_1 x_i $$

**Residual varians:**
$$ s_r^2 = \frac{SS_{res}}{df} = \frac{\sum(y_i - \hat{y}_i)^2}{n-2} $$

**Standard error af hældning:**
$$ SE(\hat{\beta}_1) = \sqrt{\frac{s_r^2}{S_{XX}}} $$

**Frihedsgrader:**
$$ df = n - 2 $$

**Teststatistik:**
$$ t = \frac{\hat{\beta}_1 - 0}{SE(\hat{\beta}_1)} = \frac{\hat{\beta}_1}{\sqrt{s_r^2 / S_{XX}}} $$

**P-værdi (tosidet):**
$$ p = 2P(T_{df} > |t|) = 2(1 - \text{CDF}_{t,df}(|t|)) $$

hvor $\text{CDF}_{t,df}$ er den kumulative fordelingsfunktion (stats.t.cdf) for t-fordelingen

```python
# T-test for regression slope: Test om der er lineær sammenhæng
# H₀: β₁ = 0 (ingen lineær sammenhæng)
# H₁: β₁ ≠ 0 (der er lineær sammenhæng)
# LaTeX: t = \frac{\hat{\beta}_1 - \beta_{1,0}}{SE(\hat{\beta}_1)} = \frac{\hat{\beta}_1}{\sqrt{s_r^2 / S_{XX}}}
# LaTeX: s_r^2 = \frac{\sum(y_i - \hat{y}_i)^2}{n-2} = \frac{SS_{res}}{n-2}
# LaTeX: SE(\hat{\beta}_1) = \sqrt{\frac{s_r^2}{S_{XX}}}, \quad df = n-2
# LaTeX: p = 2P(T_{n-2} > |t|) \text{ (tosidet)}

# Beregn middelværdier og S_XX
x_bar = np.mean(x)
y_bar = np.mean(y)
n = len(x)

# S_XX = Σ(xᵢ-x̄)²
S_XX = np.sum((x - x_bar) ** 2)

# S_XY = Σ(xᵢ-x̄)(yᵢ-ȳ)
S_XY = np.sum((x - x_bar) * (y - y_bar))

# Beregn β̂₁ (hældning)
beta_1_hat = S_XY / S_XX

# Beregn β̂₀ (skæring)
beta_0_hat = y_bar - beta_1_hat * x_bar

# Fitted values: ŷ = β̂₀ + β̂₁x
y_hat = beta_0_hat + beta_1_hat * x

# Residualer: eᵢ = yᵢ - ŷᵢ
residuals = y - y_hat

# Residual sum of squares: SS_res = Σ(yᵢ - ŷᵢ)²
SS_res = np.sum(residuals ** 2)

# Residual variance: s_r² = SS_res / (n-2)
df = n - 2  # Frihedsgrader for regression
s_r_squared = SS_res / df
s_r = np.sqrt(s_r_squared)

# Standard error af β̂₁: SE(β̂₁) = √(s_r²/S_XX)
SE_beta_1 = np.sqrt(s_r_squared / S_XX)

# Teststatistik: t = (β̂₁ - 0) / SE(β̂₁)
beta_1_0 = 0  # Nulhypotese: β₁ = 0
t_stat = (beta_1_hat - beta_1_0) / SE_beta_1

# p-værdi (tosidet test)
p_value = 2 * (1 - stats.t.cdf(abs(t_stat), df))

# Kritisk værdi
alpha = 0.05
t_critical = stats.t.ppf(1 - alpha/2, df)

print('=== T-test for lineær sammenhæng ===')
print(f'H₀: β₁ = {beta_1_0}')
print(f'H₁: β₁ ≠ {beta_1_0}')
print(f'')
print(f'β̂₁ (hældning) = {beta_1_hat:.6f}')
print(f'SE(β̂₁) = {SE_beta_1:.6f}')
print(f's_r² (residual varians) = {s_r_squared:.6f}')
print(f'S_XX = {S_XX:.4f}')
print(f'')
print(f't({df}) = {t_stat:.4f}')
print(f'Kritisk værdi (α={alpha}): ±{t_critical:.4f}')
print(f'p-værdi = {p_value:.6f}')
print(f'')
conclusion = 'Forkast H₀' if abs(t_stat) > t_critical else 'Forkast ikke H₀'
print(f'Konklusion: {conclusion}')
if abs(t_stat) > t_critical:
    print(f'Der er signifikant lineær sammenhæng mellem X og Y (p={p_value:.6f})')
else:
    print(f'Ingen signifikant lineær sammenhæng (p={p_value:.6f})')
```

### B): Bestem fordelingsfunktionen 𝐹𝑋(𝑥).

$$ F_X(x) = P(X \leq x) = \int_{-\infty}^{x} f_X(t) \, dt $$

hvor f_X(t)$ er tæthedsfunktionen (PDF)


$$ F_X(x) = P(X \leq x) = \int_{0}^{x} 3t^2 \, dt $$

$$ F_X(x) = 3 [\frac{1}{3}t^3+k]_0^x  $$

$$ F_X(x) = [t^3+k]_0^x  $$

$$ F_X(x) = (x^3+k)-(0^3+k)  $$

$$ F_X(x) = x^3  $$

brgrænsning

```python
# Bestem fordelingsfunktionen F_X(x) fra tæthedsfunktionen f_X(x)
# DANSK: F_X(x) = P(X ≤ x) = ∫_{-∞}^x f_X(t) dt
# ENGLISH: F_X(x) = P(X ≤ x) = ∫_{-∞}^x f_X(t) dt
# LaTeX: F_X(x) = P(X \leq x) = \int_{-\infty}^{x} f_X(t) \, dt
from scipy import integrate

# Definer tæthedsfunktion f_X(x)
def f_X(x):
    return 3 * x**2  # Eksempel: f(x) = 3x² for x ∈ [0,1]

# Definer fordelingsfunktion F_X(x)
def F_X(x):
    """Beregn P(X ≤ x) = ∫_{-∞}^x f_X(t) dt"""
    lower_limit = 0  # Nedre grænse for support (hvor pdf > 0)
    if x < lower_limit:
        return 0
    result, _ = integrate.quad(f_X, lower_limit, x)
    return result

# Beregn fordelingsfunktionen for specifikke værdier
x_values = [0.25, 0.5, 0.75, 1.0]
for x in x_values:
    print(f"F_X({x}) = P(X ≤ {x}) = {F_X(x):.4f}")
```

### C): Bestem den inverse CDF 𝐹𝑋−1(𝑈), som bruges i invers transformationsmetoden

**Invers fordelingsfunktion (Kvantil-funktion):**

**Definition:**

$F_X^{-1}(p)$ returnerer den $x$-værdi hvor den kumulative sandsynlighed er $p$:

$$ F_X(x) = p \quad \Rightarrow \quad x = F_X^{-1}(p) $$

**Område/Definitionsmængde:**
- Input til $F_X(x)$: $x$ værdier (reelle tal)
- Output fra $F_X(x)$: sandsynligheder $\in [0,1]$
- Input til $F_X^{-1}(p)$: sandsynligheder $p \in [0,1]$
- Output fra $F_X^{-1}(p)$: $x$ værdier

**Eksempel:**

Hvis $F_X(5) = 0.9$, så er $F_X^{-1}(0.9) = 5$

**Vigtige forholdsregler:**

1. **Strengt voksende CDF:** En ægte invers eksisterer kun hvis CDF er strengt voksende

2. **Diskrete fordelinger:** For "flade" dele defineres inversen som:
   $$ F_X^{-1}(p) = \inf\{x : F_X(x) \geq p\} $$
   (den mindste $x$ hvor $F_X(x) \geq p$)

3. **Input-begrænsning:** Brug aldrig værdier uden for $[0,1]$

4. **Grænsecases:** For kontinuerte fordelinger kan $F_X^{-1}(0)$ og $F_X^{-1}(1)$ give $-\infty$ eller $+\infty$





```python
# Beregn invers fordelingsfunktion F_X^{-1}(p) (kvantil-funktion)
# DANSK: Find x sådan at F_X(x) = P(X ≤ x) = p
# ENGLISH: Find x such that F_X(x) = P(X ≤ x) = p
# LaTeX: F_X^{-1}(p) = \inf\{x \in \mathbb{R} : F_X(x) \geq p}
from scipy import integrate, optimize

# Definer tæthedsfunktion f_X(x)
def f_X(x):
    return 3 * x**2  # Eksempel: f(x) = 3x² for x ∈ [0,1]

# Definer fordelingsfunktion F_X(x)
def F_X(x):
    """Beregn P(X ≤ x) = ∫_{-∞}^x f_X(t) dt"""
    lower_limit = 0  # Nedre grænse for support
    if x < lower_limit:
        return 0
    result, _ = integrate.quad(f_X, lower_limit, x)
    return result

# Definer invers fordelingsfunktion (kvantil-funktion)
def F_X_inv(p):
    """Find x sådan at F_X(x) = p"""
    # Løs ligningen F_X(x) - p = 0
    lower_limit = 0
    upper_limit = 1
    result = optimize.brentq(lambda x: F_X(x) - p, lower_limit, upper_limit)
    return result

# Beregn kvantiler for specifikke sandsynligheder
probabilities = [0.25, 0.5, 0.75, 0.95]
for p in probabilities:
    x = F_X_inv(p)
    print(f"F_X^{{-1}}({p}) = {x:.4f} (dvs. P(X ≤ {x:.4f}) = {p})")
```

**Kumulativ fordelingsfunktion (CDF) for diskret stokastisk variabel:**

$$ F_X(x) = P(X \leq x) = \sum_{t \leq x} f_X(t) $$

hvor $f_X(t) = P(X = t)$ er sandsynlighedsmassefunktionen (PMF)

**Egenskaber:**

1. $0 \leq F_X(x) \leq 1$ for alle $x$

2. $F_X(x)$ er ikke-aftagende (trappefunktion)

3. $\lim_{x \to -\infty} F_X(x) = 0$ og $\lim_{x \to \infty} F_X(x) = 1$

```python
# Beregn kumulativ fordelingsfunktion (CDF) fra PMF
# DANSK: F_X(x) = P(X ≤ x) = Σ_{t≤x} f_X(t)
# ENGLISH: F_X(x) = P(X ≤ x) = Σ_{t≤x} f_X(t)
# LaTeX: F_X(x) = P(X \leq x) = \sum_{t \leq x} f_X(t)


# Normaliser PMF (så summen = 1)
f_X_norm = f_X / f_X.sum()

# Sorter værdier (vigtigt for CDF)
sort_idx = np.argsort(X)
X_sorted = X[sort_idx]
f_X_sorted = f_X_norm[sort_idx]

# Beregn CDF (kumulativ sum)
F_X = np.cumsum(f_X_sorted)

# Print resultater
print("PMF (normaliseret):")
for x, fx in zip(X_sorted, f_X_sorted):
    print(f"  f_X({x}) = P(X = {x}) = {fx:.4f}")

print("\nCDF:")
for x, Fx in zip(X_sorted, F_X):
    print(f"  F_X({x}) = P(X ≤ {x}) = {Fx:.4f}")
```

En simultan tæthedsfunktion 𝑓𝑋,𝑌(𝑥, 𝑦), for to stokastiske variable 𝑋 og 𝑌 er givet ved

```python
# Definer fælles PMF matrix og værdier
# DANSK: Opret funktion til at hente P(xl ≤ X ≤ xh, yl ≤ Y ≤ yh)
# ENGLISH: Create function to query P(xl ≤ X ≤ xh, yl ≤ Y ≤ yh)

# Fælles PMF matrix (rows = X værdier, columns = Y værdier)
f_XY_raw = np.array([
    [np.nan, 0, 0.2],
    [0, 0.1, 0],
    [0.2, 0, 0.1]
])

# Definer mulige værdier for X og Y
x_vals = [-10, -15, -20]  # X værdier (skal matche antal rækker)
y_vals = [10, 15, 20]     # Y værdier (skal matche antal kolonner)

# Funktion til at hente sub-matrix baseret på værdier
def f_XY(xl, xh, yl, yh):
    """
    Hent P(xl ≤ X ≤ xh, yl ≤ Y ≤ yh)
    
    Parameters:
    xl, xh: Lower og upper bound for X
    yl, yh: Lower og upper bound for Y
    
    Returns:
    Sub-matrix af f_XY_raw
    """
    xi1 = x_vals.index(xl)
    xi2 = x_vals.index(xh)
    yi1 = y_vals.index(yl)
    yi2 = y_vals.index(yh)
    
    return f_XY_raw[
        min(xi1, xi2):max(xi1, xi2)+1,
        min(yi1, yi2):max(yi1, yi2)+1
    ]
```

### C): Opstil udtryk for kovarians 𝐶𝑜𝑣(𝑋, 𝑌) og beregn denne. Fortolk resultatet

$$ \text{Cov}(X,Y) = E[XY] - E[X]E[Y] = \int\int xy \cdot f_{X,Y}(x,y) dx dy - E[X]E[Y] $$

```python
# Beregn Cov(X,Y) fra fælles tæthedsfunktion f_{X,Y}(x,y)
# DANSK: Cov(X,Y) = E[XY] - E[X]E[Y]
# ENGLISH: Cov(X,Y) = E[XY] - E[X]E[Y]
# LaTeX: \text{Cov}(X,Y) = E[XY] - E[X]E[Y] = \int\int xy \cdot f_{X,Y}(x,y) dx dy - E[X]E[Y]
from sympy import symbols, integrate, simplify, lambdify, oo, latex
import numpy as np

# Definer symbolske variable
x, y = symbols('x y', real=True)

print("Fælles tæthedsfunktion:")
print(f"  f_{{X,Y}}(x,y) = {f_XY}")
print(fr"  LaTeX: f_{{{{X,Y}}}}(x,y) = {latex(f_XY)}")
print(f"  Support: x ∈ [{x_min}, {x_max}], y ∈ [{y_min}, {y_max}]")

# Beregn marginal tæthed for X: f_X(x) = ∫ f_{X,Y}(x,y) dy
print("\n" + "="*50)
print("MARGINAL TÆTHEDER")
print("="*50)

f_X = integrate(f_XY, (y, y_min, y_max))
f_X_simplified = simplify(f_X)
print(f"\nf_X(x) = {f_X_simplified}")
print(fr"LaTeX: f_X(x) = {latex(f_X_simplified)}")

# Beregn marginal tæthed for Y: f_Y(y) = ∫ f_{X,Y}(x,y) dx
f_Y = integrate(f_XY, (x, x_min, x_max))
f_Y_simplified = simplify(f_Y)
print(f"\nf_Y(y) = {f_Y_simplified}")
print(fr"LaTeX: f_Y(y) = {latex(f_Y_simplified)}")

# Beregn E[X] = ∫ x·f_X(x) dx
print("\n" + "="*50)
print("FORVENTNINGSVÆRDIER")
print("="*50)

E_X = integrate(x * f_X_simplified, (x, x_min, x_max))
E_X_simplified = simplify(E_X)
print(f"\nE[X] = {E_X_simplified}")
print(fr"LaTeX: E[X] = {latex(E_X_simplified)}")

# Beregn E[Y] = ∫ y·f_Y(y) dy
E_Y = integrate(y * f_Y_simplified, (y, y_min, y_max))
E_Y_simplified = simplify(E_Y)
print(f"\nE[Y] = {E_Y_simplified}")
print(fr"LaTeX: E[Y] = {latex(E_Y_simplified)}")

# Beregn E[XY] = ∫∫ xy·f_{X,Y}(x,y) dx dy
print("\n" + "="*50)
print("E[XY] - DOBBELT INTEGRAL")
print("="*50)
print(fr"\nBeregner: E[XY] = \int_{{{x_min}}}^{{{x_max}}} \int_{{{y_min}}}^{{{y_max}}} xy \cdot f_{{{{X,Y}}}}(x,y) dy dx")

E_XY = integrate(integrate(x * y * f_XY, (y, y_min, y_max)), (x, x_min, x_max))
E_XY_simplified = simplify(E_XY)

print(f"\nResultat: E[XY] = {E_XY_simplified}")
print(fr"LaTeX: E[XY] = {latex(E_XY_simplified)}")

# Beregn kovarians: Cov(X,Y) = E[XY] - E[X]E[Y]
print("\n" + "="*50)
print("KOVARIANS")
print("="*50)

Cov_XY = E_XY_simplified - E_X_simplified * E_Y_simplified
Cov_XY_simplified = simplify(Cov_XY)

print(f"\nCov(X,Y) = E[XY] - E[X]E[Y]")
print(f"Cov(X,Y) = {Cov_XY_simplified}")
print(fr"LaTeX: \text{{Cov}}(X,Y) = {latex(Cov_XY_simplified)}")

# Beregn varianserne for korrelation
print("\n" + "="*50)
print("VARIANSER")
print("="*50)

# Var(X) = E[X²] - E[X]²
E_X2 = integrate(x**2 * f_X_simplified, (x, x_min, x_max))
E_X2_simplified = simplify(E_X2)
Var_X = simplify(E_X2_simplified - E_X_simplified**2)

print(f"\nVar(X) = E[X²] - E[X]² = {Var_X}")
print(fr"LaTeX: \text{{Var}}(X) = {latex(Var_X)}")

# Var(Y) = E[Y²] - E[Y]²
E_Y2 = integrate(y**2 * f_Y_simplified, (y, y_min, y_max))
E_Y2_simplified = simplify(E_Y2)
Var_Y = simplify(E_Y2_simplified - E_Y_simplified**2)

print(f"\nVar(Y) = E[Y²] - E[Y]² = {Var_Y}")
print(fr"LaTeX: \text{{Var}}(Y) = {latex(Var_Y)}")

# Beregn korrelation: ρ = Cov(X,Y) / √(Var(X)Var(Y))
print("\n" + "="*50)
print("KORRELATION")
print("="*50)

from sympy import sqrt
Corr_XY = Cov_XY_simplified / sqrt(Var_X * Var_Y)
Corr_XY_simplified = simplify(Corr_XY)

print(f"\nρ(X,Y) = Cov(X,Y) / √(Var(X)Var(Y))")
print(f"ρ(X,Y) = {Corr_XY_simplified}")
print(fr"LaTeX: \rho_{{{{X,Y}}}} = {latex(Corr_XY_simplified)}")

# Opsummering
print("\n" + "="*50)
print("OPSUMMERING")
print("="*50)
print(f"\nE[X] = {E_X_simplified}")
print(f"E[Y] = {E_Y_simplified}")
print(f"E[XY] = {E_XY_simplified}")
print(f"Cov(X,Y) = {Cov_XY_simplified}")
print(f"Var(X) = {Var_X}")
print(f"Var(Y) = {Var_Y}")
print(f"ρ(X,Y) = {Corr_XY_simplified}")

# Numeriske værdier
try:
    print(f"\nNumeriske værdier:")
    print(f"  Cov(X,Y) ≈ {float(Cov_XY_simplified):.4f}")
    print(f"  ρ(X,Y) ≈ {float(Corr_XY_simplified):.4f}")
    
    # Fortolkning
    cov_val = float(Cov_XY_simplified)
    print(f"\nFortolkning:")
    if cov_val > 0.01:
        print("  ✓ Positiv kovarians: X og Y varierer positivt sammen")
    elif cov_val < -0.01:
        print("  ✓ Negativ kovarians: X og Y varierer negativt sammen")
    else:
        print("  ✓ Kovarians ≈ 0: Ingen lineær sammenhæng")
except:
    print("\nKan ikke beregne numeriske værdier (symbolsk udtryk indeholder variable)")
```

```python
# Beregn PDF for transformeret stokastisk variabel: Y = g(X)
# DANSK: Find f_Y(y) når Y = g(X) og f_X(x) er kendt
# ENGLISH: Find f_Y(y) when Y = g(X) and f_X(x) is known
# LaTeX: f_Y(y) = f_X(g^{-1}(y)) \cdot \left| \frac{d}{dy} g^{-1}(y) \right|
from sympy import symbols, diff, simplify, latex, solve, Abs

# Definer symbolske variable
x, y = symbols('x y', real=True)

# Definer PDF for X
f_X = (x - 1)/2  # Eksempel: f_X(x) = (x-1)/2 for x ∈ [1,3]
x_min, x_max = 1, 3  # Support for X

# Definer transformationen Y = g(X)
g = x - 4  # Eksempel: Y = X - 4 (forskydning)

print("="*70)
print("TRANSFORMATION AF STOKASTISK VARIABEL")
print("="*70)

print(f"\nPDF for X: f_X(x) = {f_X}")
print(fr"LaTeX: f_X(x) = {latex(f_X)}")
print(fr"$$ f_X(x) = {latex(f_X)} $$")
print(f"Support: x ∈ [{x_min}, {x_max}]")

print(f"\nTransformation: Y = g(X) = {g}")
print(fr"LaTeX: Y = {latex(g)}")
print(fr"$$ Y = {latex(g)} $$")

# --- TRIN 1: Find den inverse funktion g^{-1}(y) ---
print("\n" + "="*70)
print("TRIN 1: FIND INVERS FUNKTION")
print("="*70)

# Løs y = g(x) for x
inverse_solutions = solve(y - g, x)
if inverse_solutions:
    g_inv = inverse_solutions[0]  # Tag den første løsning
    print(f"\nLøser: y = {g} for x")
    print(f"Invers funktion: x = g⁻¹(y) = {g_inv}")
    print(fr"LaTeX: x = {latex(g_inv)}")
    print(fr"$$ x = g^{{-1}}(y) = {latex(g_inv)} $$")
else:
    print("\n⚠ Kunne ikke finde invers funktion!")
    g_inv = None

if g_inv is not None:
    # --- TRIN 2: Beregn Jacobi-determinanten ---
    print("\n" + "="*70)
    print("TRIN 2: BEREGN JACOBI-DETERMINANT")
    print("="*70)
    
    # Differentier g^{-1}(y) med hensyn til y
    jacobian = diff(g_inv, y)
    jacobian_simplified = simplify(jacobian)
    jacobian_abs = Abs(jacobian_simplified)
    
    print(f"\nDifferentier: d/dy[g⁻¹(y)] = d/dy[{g_inv}]")
    print(f"Resultat: {jacobian_simplified}")
    print(fr"LaTeX: \frac{{d}}{{dy}} g^{{-1}}(y) = {latex(jacobian_simplified)}")
    print(fr"$$ \frac{{d}}{{dy}} g^{{-1}}(y) = {latex(jacobian_simplified)} $$")
    
    print(f"\nNumerisk værdi: |d/dy g⁻¹(y)| = {jacobian_abs}")
    print(fr"$$ \left| \frac{{d}}{{dy}} g^{{-1}}(y) \right| = {latex(jacobian_abs)} $$")
    
    # --- TRIN 3: Anvend transformationsformlen ---
    print("\n" + "="*70)
    print("TRIN 3: ANVEND TRANSFORMATIONSFORMEL")
    print("="*70)
    
    print("\nFormel: f_Y(y) = f_X(g⁻¹(y)) · |d/dy g⁻¹(y)|")
    print(fr"$$ f_Y(y) = f_X(g^{{-1}}(y)) \cdot \left| \frac{{d}}{{dy}} g^{{-1}}(y) \right| $$")
    
    # Substituer x = g^{-1}(y) i f_X(x)
    f_Y = f_X.subs(x, g_inv) * jacobian_abs
    f_Y_simplified = simplify(f_Y)
    
    print(f"\nSubstituer x = {g_inv} i f_X(x) = {f_X}")
    print(f"f_Y(y) = {f_Y_simplified}")
    print(fr"LaTeX: f_Y(y) = {latex(f_Y_simplified)}")
    print(fr"$$ f_Y(y) = {latex(f_Y_simplified)} $$")
    
    # --- TRIN 4: Bestem support for Y ---
    print("\n" + "="*70)
    print("TRIN 4: BESTEM SUPPORT/INTERVAL FOR Y")
    print("="*70)
    
    print("\nMetode: Beregn y_min = g(x_min) og y_max = g(x_max)")
    print("Tjek om g er voksende eller aftagende for at finde det korrekte interval")
    
    # Beregn y_min og y_max
    y_at_xmin = g.subs(x, x_min)
    y_at_xmax = g.subs(x, x_max)
    
    print(f"\ny ved nedre grænse: g({x_min}) = {y_at_xmin}")
    print(f"y ved øvre grænse: g({x_max}) = {y_at_xmax}")
    
    # Check om g er voksende eller aftagende
    diff_check = simplify(y_at_xmax - y_at_xmin)
    
    if diff_check > 0:
        y_min_val, y_max_val = y_at_xmin, y_at_xmax
        print(f"\ng er VOKSENDE (g(x_max) > g(x_min))")
        print(f"Derfor: Y ∈ [g(x_min), g(x_max)] = [{y_min_val}, {y_max_val}]")
    elif diff_check < 0:
        y_min_val, y_max_val = y_at_xmax, y_at_xmin
        print(f"\ng er AFTAGENDE (g(x_max) < g(x_min))")
        print(f"Derfor: Y ∈ [g(x_max), g(x_min)] = [{y_min_val}, {y_max_val}] (omvendt!)")
    else:
        y_min_val, y_max_val = y_at_xmin, y_at_xmax
        print(f"\ng er KONSTANT eller ubestemmelig")
    
    print(f"\n→ X-support: x ∈ [{x_min}, {x_max}]")
    print(f"→ Y-support: y ∈ [{y_min_val}, {y_max_val}]")
    print(fr"$$ X \in [{latex(x_min)}, {latex(x_max)}] \quad \Rightarrow \quad Y \in [{latex(y_min_val)}, {latex(y_max_val)}] $$")
    
    # --- OPSUMMERING ---
    print("\n" + "="*70)
    print("OPSUMMERING / SUMMARY")
    print("="*70)
    
    print(f"\nOriginal variabel X:")
    print(f"  PDF: f_X(x) = {f_X}")
    print(f"  Support: x ∈ [{x_min}, {x_max}]")
    
    print(f"\nTransformation: Y = {g}")
    
    print(f"\nTransformeret variabel Y:")
    print(f"  PDF: f_Y(y) = {f_Y_simplified}")
    print(f"  Support: y ∈ [{y_min_val}, {y_max_val}]")
    
    print(f"\nLaTeX:")
    print(fr"$$ f_Y(y) = {latex(f_Y_simplified)}, \quad y \in [{latex(y_min_val)}, {latex(y_max_val)}] $$")
```
