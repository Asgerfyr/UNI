For this, the formula for total probability is used. This is evaluated by taking the probability that you have the bacteria times the probability of the test being positive with the bacteria, added with the probability that you don't have the bacteria times the probability that the test results in positive without the bacteria.

```python
# Total sandsynlighed med to komplementære hændelser
# P(B) = P(A) × P(B|A) + P(A^c) × P(B|A^c)
# P(B) = P(A) × P(B|A) + (1-P(A)) × P(B|A^c)

P_B = P_B_T  # P(A) - sandsynlighed for hændelse A
P_P_B = P_P_B  # P(B|A)
P_P_nB = P_P_nB  # P(B|A^c)

# Total sandsynlighed
P_P = P_B * P_P_B + (1 - P_B) * P_P_nB
print(f"P(P) = {P_P:.4f}")
print(round(P_P*100, 2), "%")
```

Vi ønsker at undersøge om de nye og gamle vindmøller producerer samme mængde energi

### A Opstil et udtryk for normalfordelingen for effektforskellen mellem de nye og gamle vindmøller

```python
import numpy as np
from scipy import stats
from scipy.stats import norm, t, chi2, f, binom, poisson, expon, uniform, bernoulli, geom
import matplotlib.pyplot as plt

gammel = np.array([11.3,17.4,12.4,14.5,23.5,19.6,10.5,40.7,35.6,42.5])
ny = np.array([12,19.7,14.8,15.4,28.2,22.2,13.5,42.7,38.6,45.8])

diff = ny-gammel

mean = np.mean(diff)

# Variance from simulated data
variance = np.sum((diff - np.mean(diff)) ** 2) / (len(diff) - 1)
print(f"Variance: {variance:.6f}")


mu = mean
sigma2 = variance

print(f"D = N(mu_diff, sigma^2_diff) = N({mu:.2f},{sigma2:.2f})")
```

### B Opskriv en nulhypotese og alternativ hypotese for testen

$ H_0: \mu_D =0 $ den nye og den gamle vindmølle generare samme mængde energi

$ H_1: \mu_D \neq 0 $ den nye og den gamle vindmølle generare ikke samme mængde energi

### C Opstil udtryk for og beregn p-værdien. Kan vi afvise nulhypotesen på 5% signifikansniveau?

```python
import numpy as np
from scipy import stats
from scipy.stats import norm, t, chi2, f, binom, poisson, expon, uniform, bernoulli, geom
import matplotlib.pyplot as plt

# t-test for middelværdi (ukendt varians)
# H₀: μ = μ₀  vs  H₁: μ ≠ μ₀
sample_mean = mean
mu_0 = 0
s = np.sqrt(variance)  # Stikprøvestandardafvigelse
n = 10

# Teststatistik: t = (x̄ - μ₀) / (s/√n)
t_stat = (sample_mean - mu_0) / (s / np.sqrt(n))
df = n - 1

alpha = 0.05
t_critical = stats.t.ppf(1 - alpha/2, df)
p_value = 2 * (1 - stats.t.cdf(abs(t_stat), df))

print(f"H₀: μ = {mu_0}")
print(f"H₁: μ ≠ {mu_0}")
print(f"t({df}) = {t_stat:.4f}")
print(f"Kritisk værdi: ±{t_critical:.4f}")
print(f"p-værdi = {p_value:.7f}")
print(f"Konklusion: {'Forkast H₀' if abs(t_stat) > t_critical else 'Forkast ikke H₀'}")
```

### D Opstil udtryk for og beregn 99% konfidensintervallet for forskellen i genereret energi mellem nye og gamle vindmøller.

```python
# Konfidensinterval for middelværdi μ (ukendt populationsvarians)
# BRUG NÅR: Populationsstandardafvigelse σ er UKENDT - bruger stikprøve-std s
# Bruger T-FORDELING med df = n-1 frihedsgrader
# Antager: Data er tilnærmelsesvis normalfordelt
# CI: x̄ ± t_{α/2,n-1} × (s/√n)
# LaTeX: CI = \left[\bar{x} - t_{\alpha/2,n-1}\frac{s}{\sqrt{n}}, \bar{x} + t_{\alpha/2,n-1}\frac{s}{\sqrt{n}}\right]
alpha = 0.01  # Signifikansniveau
sample_mean = mean  # x̄ - Observeret stikprøvemiddelværdi
s = s  # s - Stikprøvestandardafvigelse (beregnet fra data)
n = 10  # Stikprøvestørrelse

df = n - 1  # Frihedsgrader
t_critical = stats.t.ppf(1 - alpha/2, df)
SE = s / np.sqrt(n)
margin_error = t_critical * SE

lower_bound = sample_mean - margin_error
upper_bound = sample_mean + margin_error

print(f"{(1-alpha)*100:.0f}% Konfidensinterval: ({lower_bound:.4f}, {upper_bound:.4f})")
print(f"Frihedsgrader: {df}")
```

Ud fra resultatet i opgave D kan der observeres, at nulhypotesen ligger uden for dette konfidensinterval, hvilket yderligere verificerer forkastelsen af nulhypotesen.

```python
# T-test for regression slope: Test om der er lineær sammenhæng
# H₀: β₁ = 0 (ingen lineær sammenhæng)
# H₁: β₁ ≠ 0 (der er lineær sammenhæng)
# LaTeX: t = \frac{\hat{\beta}_1 - \beta_{1,0}}{SE(\hat{\beta}_1)} = \frac{\hat{\beta}_1}{\sqrt{s_r^2 / S_{XX}}}
# LaTeX: s_r^2 = \frac{\sum(y_i - \hat{y}_i)^2}{n-2} = \frac{SS_{res}}{n-2}
# LaTeX: SE(\hat{\beta}_1) = \sqrt{\frac{s_r^2}{S_{XX}}}, \quad df = n-2
# LaTeX: p = 2P(T_{n-2} > |t|) \text{ (tosidet)}


# Beregn middelværdier og S_XX
x_bar = np.mean(x)
y_bar = np.mean(y)
n = len(x)

# S_XX = Σ(xᵢ-x̄)²
S_XX = np.sum((x - x_bar) ** 2)

# S_XY = Σ(xᵢ-x̄)(yᵢ-ȳ)
S_XY = np.sum((x - x_bar) * (y - y_bar))

# Beregn β̂₁ (hældning)
beta_1_hat = S_XY / S_XX

# Beregn β̂₀ (skæring)
beta_0_hat = y_bar - beta_1_hat * x_bar

# Fitted values: ŷ = β̂₀ + β̂₁x
y_hat = beta_0_hat + beta_1_hat * x

# Residualer: eᵢ = yᵢ - ŷᵢ
residuals = y - y_hat

# Residual sum of squares: SS_res = Σ(yᵢ - ŷᵢ)²
SS_res = np.sum(residuals ** 2)

# Residual variance: s_r² = SS_res / (n-2)
df = n - 2  # Frihedsgrader for regression
s_r_squared = SS_res / df
s_r = np.sqrt(s_r_squared)

# Standard error af β̂₁: SE(β̂₁) = √(s_r²/S_XX)
SE_beta_1 = np.sqrt(s_r_squared / S_XX)

# Teststatistik: t = (β̂₁ - 0) / SE(β̂₁)
beta_1_0 = 0  # Nulhypotese: β₁ = 0
t_stat = (beta_1_hat - beta_1_0) / SE_beta_1

# p-værdi (tosidet test)
p_value = 2 * (1 - stats.t.cdf(abs(t_stat), df))

# Kritisk værdi
alpha = 0.05
t_critical = stats.t.ppf(1 - alpha/2, df)

print('=== T-test for lineær sammenhæng ===')
print(f'H₀: β₁ = {beta_1_0}')
print(f'H₁: β₁ ≠ {beta_1_0}')
print(f'')
print(f'β̂₁ (hældning) = {beta_1_hat:.6f}')
print(f'SE(β̂₁) = {SE_beta_1:.6f}')
print(f's_r² (residual varians) = {s_r_squared:.6f}')
print(f'S_XX = {S_XX:.4f}')
print(f'')
print(f't({df}) = {t_stat:.4f}')
print(f'Kritisk værdi (α={alpha}): ±{t_critical:.4f}')
print(f'p-værdi = {p_value}')
print(f'')
conclusion = 'Forkast H₀' if abs(t_stat) > t_critical else 'Forkast ikke H₀'
print(f'Konklusion: {conclusion}')
if abs(t_stat) > t_critical:
    print(f'Der er signifikant lineær sammenhæng mellem X og Y (p={p_value:.6f})')
else:
    print(f'Ingen signifikant lineær sammenhæng (p={p_value:.6f})')
```

### E Antag at p=0.18 og benyt CLT til at beregne sandsynligheden for at finde mere end 54 fej

**Centralgrænseværdisætningen (CLT):**

Givet X_1, X_2, \ldots, X_n$ uafhængige med E[X_i] = \mu$ og $\text{Var}(X_i) = \sigma^2$

**Stikprøvemiddelværdi:**
$$ \bar{X} = \frac{1}{n}\sum_{i=1}^{n} X_i \xrightarrow{d} N\left(\mu, \frac{\sigma^2}{n}\right) \text{ når } n \to \infty $$

**Binomial approximation:**
$$ X \sim \text{Binomial}(n, p) \approx N(np, np(1-p)) $$
når np \geq 5$ og n(1-p) \geq 5$

**Proportion approximation:**
$$ \hat{p} = \frac{X}{n} \approx N\left(p, \frac{p(1-p)}{n}\right) $$

**Standardisering:**
$$ Z = \frac{\bar{X} - \mu}{\sigma/\sqrt{n}} \xrightarrow{d} N(0,1) $$

```python
import numpy as np
from scipy import stats
from scipy.stats import norm, t, chi2, f, binom, poisson, expon, uniform, bernoulli, geom
import matplotlib.pyplot as plt

# Central Limit Theorem: Binomial → Normal approximation
# X ~ Binomial(n, p) ≈ N(np, np(1-p)) når n er stor
# Regel: Brug når np ≥ 5 og n(1-p) ≥ 5

n = 3000  # Antal forsøg
p = 0.018  # Succesandsynlighed

# Check CLT betingelser
print(f"np = {n*p:.1f}, n(1-p) = {n*(1-p):.1f}")
print(f"CLT gyldig: {n*p >= 5 and n*(1-p) >= 5}")

# Binomial parametre → Normal parametre
mu = n * p
sigma = np.sqrt(n * p * (1 - p))

print(f"X ~ Binomial({n}, {p}) ≈ N({mu:.2f}, {sigma**2:.2f})")
print(f"μ = {mu:.2f}, σ = {sigma:.4f}")

# Beregn sandsynlighed: P(X > k) med kontinuitetskorrektion
k = 54  # Kritisk værdi

# Med kontinuitetskorrektion: P(X > k) = P(X ≥ k+1) ≈ P(Z > (k+0.5-μ)/σ)
z = (k + 0.5 - mu) / sigma
p_greater = 1 - stats.norm.cdf(z)

print(f"")
print(f"P(X > {k}) med kontinuitetskorrektion:")
print(f"z = ({k}+0.5-{mu:.2f})/{sigma:.4f} = {z:.4f}")
print(f"P(X > {k}) = P(Z > {z:.4f}) = {p_greater:.6f}")
```

### C): Opstil et udtryk for den poolede varians og beregn denne

$$ s_p^2 = \frac{(n_1-1)s_1^2+(n_2-1)s_2^2}{n_1+n_2-2} $$

$$ s_1^2 = \frac{1}{n_1-1}\sum_{i=1}^{n_1}(x_{1i}-\bar{x}_1)^2 $$
$$ s_2^2 = \frac{1}{n_2-1}\sum_{i=1}^{n_2}(x_{2i}-\bar{x}_2)^2 $$

$$ s_p^2 = \frac{\sum_{i=1}^{n_1}(x_{1i}-\bar{x}_1)^2+\sum_{i=1}^{n_2}(x_{2i}-\bar{x}_2)^2}{n_1+n_2-2} $$

```python

n1 = n_ta
s1_sq = np.sum((ta - np.mean(ta))**2)/(n1-1)


n2 = n_ny
s2_sq = np.sum((ny - np.mean(ny))**2)/(n2-1)



# Pooled variance for two samples
# LaTeX: s_p^2 = \frac{(n_1-1)s_1^2 + (n_2-1)s_2^2}{n_1+n_2-2}
s_pooled_sq = ((n1 - 1) * s1_sq + (n2 - 1) * s2_sq) / (n1 + n2 - 2)
s_pooled = np.sqrt(s_pooled_sq)
print(f"Pooled variance: {s_pooled_sq:.4f}")
print(f"Pooled std: {s_pooled:.4f}")
```

```python
# P-værdi beregning for to-stikprøve t-test med pooled varians
# LaTeX: t = \frac{(\bar{x}_1 - \bar{x}_2) - (\mu_1 - \mu_2)}{s_p\sqrt{\frac{1}{n_1} + \frac{1}{n_2}}}
# LaTeX: p = 2P(T_{df} > |t|) = 2(1 - \text{CDF}_{t,df}(|t|))

# Data fra to grupper
x1 = ny
x2 = ta

# Beskrivende statistik
n1, n2 = len(x1), len(x2)
x1_bar = np.mean(x1)
x2_bar = np.mean(x2)


# Pooled varians: s_p^2
s_p_sq = s_pooled_sq
s_p = s_pooled

# Teststatistik (generelt): t = ((x̄₁ - x̄₂) - (μ₁ - μ₂)) / (s_p√(1/n₁ + 1/n₂))
mu_diff = 0  # (μ₁ - μ₂) under H₀, typisk 0
t_stat = ((x1_bar - x2_bar) - mu_diff) / (s_p * np.sqrt(1/n1 + 1/n2))

# Frihedsgrader
df = n1 + n2 - 2

# P-værdi (tosidet): p = 2P(T_df > |t|) = 2(1 - CDF_t(|t|))
p_value = 1 * (1 - stats.t.cdf(abs(t_stat), df))

print(f"x̄₁ = {x1_bar:.4f}, x̄₂ = {x2_bar:.4f}")
print(f"s_p² = {s_p_sq:.4f}, s_p = {s_p:.4f}")
print(f"t({df}) = {t_stat:.4f}")
print(f"p-værdi = {p_value}")
```

```python
n_kald = np.array([10, 20, 30, 40, 50])
ms_responce = np.array([47, 55, 63, 72, 81])
```

### A): Antag en lineær regressionsmodel, og angiv modellen på symbolfor

$$ \hat{\beta}_1 = \frac{S_{XY}}{S_{XX}} = \frac{\sum(x_i-\bar{x})(y_i-\bar{y})}{\sum(x_i-\bar{x})^2} $$

$$ \hat{\beta}_0 = \bar{y} - \hat{\beta}_1\bar{x} $$

$$ y=\hat{\beta}_0+\hat{\beta}_1x $$

### B): Beregn modellens parametre (hældning og skæring med y-aksen), og angiv hvordan disse er beregne

$$ \hat{\beta}_1 = \frac{S_{XY}}{S_{XX}} = \frac{\sum(x_i-\bar{x})(y_i-\bar{y})}{\sum(x_i-\bar{x})^2} $$
$$ \hat{\beta}_0 = \bar{y} - \hat{\beta}_1\bar{x} $$
$$ R^2 = \frac{S_{XY}^2}{S_{XX} \cdot S_{YY}} = 1 - \frac{\sum(y_i-\hat{y}_i)^2}{\sum(y_i-\bar{y})^2} $$

```python
# Lineær regression - manuel beregning med formler
# Model: Y = β₀ + β₁X + ε where ε ~ N(0,σ²)
# LaTeX: \hat{\beta}_1 = \frac{S_{XY}}{S_{XX}} = \frac{\sum(x_i-\bar{x})(y_i-\bar{y})}{\sum(x_i-\bar{x})^2}
# LaTeX: \hat{\beta}_0 = \bar{y} - \hat{\beta}_1\bar{x}
# LaTeX: R^2 = \frac{S_{XY}^2}{S_{XX} \cdot S_{YY}} = 1 - \frac{\sum(y_i-\hat{y}_i)^2}{\sum(y_i-\bar{y})^2}
x = n_kald
y = ms_responce

# Beregn middelværdier
x_bar = np.mean(x)
y_bar = np.mean(y)

# Beregn sum of squares og cross-products
# S_XX = Σ(xᵢ-x̄)²
S_XX = np.sum((x - x_bar) ** 2)

# S_YY = Σ(yᵢ-ȳ)²
S_YY = np.sum((y - y_bar) ** 2)

# S_XY = Σ(xᵢ-x̄)(yᵢ-ȳ)
S_XY = np.sum((x - x_bar) * (y - y_bar))

# Beregn regressionskoefficienter
# β̂₁ = S_XY / S_XX (hældning/slope)
beta_1 = S_XY / S_XX

# β̂₀ = ȳ - β̂₁·x̄ (skæring/intercept)
beta_0 = y_bar - beta_1 * x_bar

# Fitted values: ŷ = β̂₀ + β̂₁x
y_hat = beta_0 + beta_1 * x

# Beregn R² (forklaringsgrad)
# Metode 1: R² = S_XY² / (S_XX · S_YY)
R_squared = (S_XY ** 2) / (S_XX * S_YY)

# Metode 2: R² = 1 - SS_res/SS_tot
SS_res = np.sum((y - y_hat) ** 2)  # Residual sum of squares
SS_tot = S_YY  # Total sum of squares
R_squared_alt = 1 - (SS_res / SS_tot)

# Regressionsligningen
print(f'Regressionsmodel: ŷ = {beta_0:.4f} + {beta_1:.4f}x')
print(f'β̂₀ (intercept) = {beta_0:.4f}')
print(f'β̂₁ (slope) = {beta_1:.4f}')
print(f'R² = {R_squared:.4f}')
print(f'')
print(f'Sum of squares:')
print(f'S_XX = {S_XX:.4f}')
print(f'S_YY = {S_YY:.4f}')
print(f'S_XY = {S_XY:.4f}')
```

### F): Udfør en hypotesetest for signifikansen af hældningskoefficiente

𝐻0: 𝛽1 = 0 og 𝐻1: 𝛽1 ≠ 0

Vurder, om der er statistisk evidens for en lineær sammenhæng mellem belastning og responstid ved et signifikansniveau på 5%.

**Hældning:**
$$ \hat{\beta}_1 = \frac{S_{XY}}{S_{XX}} = \frac{\sum(x_i-\bar{x})(y_i-\bar{y})}{\sum(x_i-\bar{x})^2} $$

**Fitted values:**
$$ \hat{y}_i = \hat{\beta}_0 + \hat{\beta}_1 x_i $$

**Residual varians:**
$$ s_r^2 = \frac{SS_{res}}{df} = \frac{\sum(y_i - \hat{y}_i)^2}{n-2} $$

**Standard error af hældning:**
$$ SE(\hat{\beta}_1) = \sqrt{\frac{s_r^2}{S_{XX}}} $$

**Frihedsgrader:**
$$ df = n - 2 $$

**Teststatistik:**
$$ t = \frac{\hat{\beta}_1 - 0}{SE(\hat{\beta}_1)} = \frac{\hat{\beta}_1}{\sqrt{s_r^2 / S_{XX}}} $$

**P-værdi (tosidet):**
$$ p = 2P(T_{df} > |t|) = 2(1 - \text{CDF}_{t,df}(|t|)) $$

hvor $\text{CDF}_{t,df}$ er den kumulative fordelingsfunktion (stats.t.cdf) for t-fordelingen

```python
# T-test for regression slope: Test om der er lineær sammenhæng
# H₀: β₁ = 0 (ingen lineær sammenhæng)
# H₁: β₁ ≠ 0 (der er lineær sammenhæng)
# LaTeX: t = \frac{\hat{\beta}_1 - \beta_{1,0}}{SE(\hat{\beta}_1)} = \frac{\hat{\beta}_1}{\sqrt{s_r^2 / S_{XX}}}
# LaTeX: s_r^2 = \frac{\sum(y_i - \hat{y}_i)^2}{n-2} = \frac{SS_{res}}{n-2}
# LaTeX: SE(\hat{\beta}_1) = \sqrt{\frac{s_r^2}{S_{XX}}}, \quad df = n-2
# LaTeX: p = 2P(T_{n-2} > |t|) \text{ (tosidet)}

# Beregn middelværdier og S_XX
x_bar = np.mean(x)
y_bar = np.mean(y)
n = len(x)

# S_XX = Σ(xᵢ-x̄)²
S_XX = np.sum((x - x_bar) ** 2)

# S_XY = Σ(xᵢ-x̄)(yᵢ-ȳ)
S_XY = np.sum((x - x_bar) * (y - y_bar))

# Beregn β̂₁ (hældning)
beta_1_hat = S_XY / S_XX

# Beregn β̂₀ (skæring)
beta_0_hat = y_bar - beta_1_hat * x_bar

# Fitted values: ŷ = β̂₀ + β̂₁x
y_hat = beta_0_hat + beta_1_hat * x

# Residualer: eᵢ = yᵢ - ŷᵢ
residuals = y - y_hat

# Residual sum of squares: SS_res = Σ(yᵢ - ŷᵢ)²
SS_res = np.sum(residuals ** 2)

# Residual variance: s_r² = SS_res / (n-2)
df = n - 2  # Frihedsgrader for regression
s_r_squared = SS_res / df
s_r = np.sqrt(s_r_squared)

# Standard error af β̂₁: SE(β̂₁) = √(s_r²/S_XX)
SE_beta_1 = np.sqrt(s_r_squared / S_XX)

# Teststatistik: t = (β̂₁ - 0) / SE(β̂₁)
beta_1_0 = 0  # Nulhypotese: β₁ = 0
t_stat = (beta_1_hat - beta_1_0) / SE_beta_1

# p-værdi (tosidet test)
p_value = 2 * (1 - stats.t.cdf(abs(t_stat), df))

# Kritisk værdi
alpha = 0.05
t_critical = stats.t.ppf(1 - alpha/2, df)

print('=== T-test for lineær sammenhæng ===')
print(f'H₀: β₁ = {beta_1_0}')
print(f'H₁: β₁ ≠ {beta_1_0}')
print(f'')
print(f'β̂₁ (hældning) = {beta_1_hat:.6f}')
print(f'SE(β̂₁) = {SE_beta_1:.6f}')
print(f's_r² (residual varians) = {s_r_squared:.6f}')
print(f'S_XX = {S_XX:.4f}')
print(f'')
print(f't({df}) = {t_stat:.4f}')
print(f'Kritisk værdi (α={alpha}): ±{t_critical:.4f}')
print(f'p-værdi = {p_value:.6f}')
print(f'')
conclusion = 'Forkast H₀' if abs(t_stat) > t_critical else 'Forkast ikke H₀'
print(f'Konklusion: {conclusion}')
if abs(t_stat) > t_critical:
    print(f'Der er signifikant lineær sammenhæng mellem X og Y (p={p_value:.6f})')
else:
    print(f'Ingen signifikant lineær sammenhæng (p={p_value:.6f})')
```

### A): Bestem forventningsværdi og varians af 𝑋

**Middelværdi for en kontinuert stokastisk variabel:**
$$ E[X] = \int_{-\infty}^{\infty} x \cdot f_X(x) \, dx $$

**Varians:**
$$ Var(X) = E[X^2] - E[X]^2 $$

hvor:
$$ E[X^2] = \int_{-\infty}^{\infty} x^2 \cdot f_X(x) \, dx $$

da det kun er i intervallet 0 til 1 kan dette forsimples

**Middelværdi for en kontinuert stokastisk variabel:**
$$ E[X] = \int_{0}^{1} x \cdot 3x^2 \, dx $$

**Varians:**
$$ Var(X) = E[X^2] - E[X]^2 $$

hvor:
$$ E[X^2] = \int_{0}^{1} x^2 \cdot 3x^2 \, dx $$

```python
# Beregn E[X] og Var(X) for kontinuert stokastisk variabel
# DANSK: Middelværdi E[X] = ∫ x·f(x) dx
# DANSK: Varians Var(X) = E[X²] - E[X]²
# ENGLISH: Expected value E[X] = ∫ x·f(x) dx
# ENGLISH: Variance Var(X) = E[X²] - E[X]²
# LaTeX: E[X] = \int_{-\infty}^{\infty} x \cdot f_X(x) \, dx
# LaTeX: E[X^2] = \int_{-\infty}^{\infty} x^2 \cdot f_X(x) \, dx
# LaTeX: Var(X) = E[X^2] - E[X]^2
from scipy import integrate

# Definer tæthedsfunktion f_X(x)
def f_X(x):
    return 3 * x**2  # Eksempel: f(x) = 3x² for x ∈ [0,1]

# Integrationsgränser
a = 0  # Nedre grænse
b = 1  # Øvre grænse

# Beregn E[X] = ∫ x·f(x) dx
E_X, _ = integrate.quad(lambda x: x * f_X(x), a, b)

# Beregn E[X²] = ∫ x²·f(x) dx
E_X2, _ = integrate.quad(lambda x: x**2 * f_X(x), a, b)

# Beregn Var(X) = E[X²] - E[X]²
Var_X = E_X2 - E_X**2

print(f"E[X] = {E_X:.4f}")
print(f"E[X²] = {E_X2:.4f}")
print(f"Var(X) = {Var_X:.4f}")
print(f"SD(X) = {np.sqrt(Var_X):.4f}")
```

### B): Bestem fordelingsfunktionen 𝐹𝑋(𝑥).

**Fordelingsfunktion F_X(x):**
$$ F_X(x) = P(X \leq x) = \int_{-\infty}^{x} f_X(t) \, dt $$

hvor f_X(t)$ er tæthedsfunktionen (PDF)


$$ F_X(x) = P(X \leq x) = \int_{0}^{x} 3t^2 \, dt $$

$$ F_X(x) = 3 \int_{0}^{x} t^2 \, dt $$

$$ F_X(x) = 3 [\frac{1}{3}t^3+k]_0^x  $$

$$ F_X(x) = [t^3+k]_0^x  $$

$$ F_X(x) = (x^3+k)-(0^3+k)  $$

$$ F_X(x) = x^3+k-k  $$

$$ F_X(x) = x^3  $$

brgrænsning

$$ F_X(x) = 
\begin{cases}
0 & \text{if } x \lt 0 \\
x^3 & \text{if } 0 \le x \le 1 \\
1 & \text{if } x \gt 1 \\
\end{cases}
$$

```python
# Bestem fordelingsfunktionen F_X(x) fra tæthedsfunktionen f_X(x)
# DANSK: F_X(x) = P(X ≤ x) = ∫_{-∞}^x f_X(t) dt
# ENGLISH: F_X(x) = P(X ≤ x) = ∫_{-∞}^x f_X(t) dt
# LaTeX: F_X(x) = P(X \leq x) = \int_{-\infty}^{x} f_X(t) \, dt
from scipy import integrate

# Definer tæthedsfunktion f_X(x)
def f_X(x):
    return 3 * x**2  # Eksempel: f(x) = 3x² for x ∈ [0,1]

# Definer fordelingsfunktion F_X(x)
def F_X(x):
    """Beregn P(X ≤ x) = ∫_{-∞}^x f_X(t) dt"""
    lower_limit = 0  # Nedre grænse for support (hvor pdf > 0)
    if x < lower_limit:
        return 0
    result, _ = integrate.quad(f_X, lower_limit, x)
    return result

# Beregn fordelingsfunktionen for specifikke værdier
x_values = [0.25, 0.5, 0.75, 1.0]
for x in x_values:
    print(f"F_X({x}) = P(X ≤ {x}) = {F_X(x):.4f}")
```

### C): Bestem den inverse CDF 𝐹𝑋−1(𝑈), som bruges i invers transformationsmetoden

**Invers fordelingsfunktion (Kvantil-funktion):**

**Definition:**

$F_X^{-1}(p)$ returnerer den $x$-værdi hvor den kumulative sandsynlighed er $p$:

$$ F_X(x) = p \quad \Rightarrow \quad x = F_X^{-1}(p) $$

**Område/Definitionsmængde:**
- Input til $F_X(x)$: $x$ værdier (reelle tal)
- Output fra $F_X(x)$: sandsynligheder $\in [0,1]$
- Input til $F_X^{-1}(p)$: sandsynligheder $p \in [0,1]$
- Output fra $F_X^{-1}(p)$: $x$ værdier

**Eksempel:**

Hvis $F_X(5) = 0.9$, så er $F_X^{-1}(0.9) = 5$

**Vigtige forholdsregler:**

1. **Strengt voksende CDF:** En ægte invers eksisterer kun hvis CDF er strengt voksende

2. **Diskrete fordelinger:** For "flade" dele defineres inversen som:
   $$ F_X^{-1}(p) = \inf\{x : F_X(x) \geq p\} $$
   (den mindste $x$ hvor $F_X(x) \geq p$)

3. **Input-begrænsning:** Brug aldrig værdier uden for $[0,1]$

4. **Grænsecases:** For kontinuerte fordelinger kan $F_X^{-1}(0)$ og $F_X^{-1}(1)$ give $-\infty$ eller $+\infty$





```python
# Beregn invers fordelingsfunktion F_X^{-1}(p) (kvantil-funktion)
# DANSK: Find x sådan at F_X(x) = P(X ≤ x) = p
# ENGLISH: Find x such that F_X(x) = P(X ≤ x) = p
# LaTeX: F_X^{-1}(p) = \inf\{x \in \mathbb{R} : F_X(x) \geq p}
from scipy import integrate, optimize

# Definer tæthedsfunktion f_X(x)
def f_X(x):
    return 3 * x**2  # Eksempel: f(x) = 3x² for x ∈ [0,1]

# Definer fordelingsfunktion F_X(x)
def F_X(x):
    """Beregn P(X ≤ x) = ∫_{-∞}^x f_X(t) dt"""
    lower_limit = 0  # Nedre grænse for support
    if x < lower_limit:
        return 0
    result, _ = integrate.quad(f_X, lower_limit, x)
    return result

# Definer invers fordelingsfunktion (kvantil-funktion)
def F_X_inv(p):
    """Find x sådan at F_X(x) = p"""
    # Løs ligningen F_X(x) - p = 0
    lower_limit = 0
    upper_limit = 1
    result = optimize.brentq(lambda x: F_X(x) - p, lower_limit, upper_limit)
    return result

# Beregn kvantiler for specifikke sandsynligheder
probabilities = [0.25, 0.5, 0.75, 0.95]
for p in probabilities:
    x = F_X_inv(p)
    print(f"F_X^{{-1}}({p}) = {x:.4f} (dvs. P(X ≤ {x:.4f}) = {p})")
```
