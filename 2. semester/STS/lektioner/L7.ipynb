E[M_n]
$$E[M_n] = \frac{E[X_1]+E[X_2]+..+E[X_n]}{n}$$
$$E[M_n] = \frac{n*\frac{1}{2}}{n} = \frac{1}{2}$$


Var[M_n]
$$Var[M_n]=\frac{1}{n^2}*\sum_{x=1}^n Var[X_n] =\frac{1}{n^2}*\sum_{x=1}^n E[X^2]-E[X]^2$$
$$E[X^2]=\int_0^1 x^2 dx = \frac{1}{3}$$
$$E[X]^2=\frac{1^2}{2^2}=\frac{1}{4}$$

$$Var[X]=\frac{1}{3}-\frac{1}{4}=\frac{1}{12}$$

$$Var[M_n]=\frac{1}{n^2}*\sum_{x=1}^n Var[X_n]=\frac{1}{n^2}*n*\frac{1}{12}=\frac{1}{12n}$$
