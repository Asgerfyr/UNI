```python
from scipy.stats import norm
import numpy as np


var_X = 81

sigma_X_squared = var_X

sigma_X = np.sqrt(sigma_X_squared)

sam_mean = 50.1


a = 0.05

z_neg_a_over_2 = norm.ppf(a/2)
z_a_over_2 = norm.ppf(1-a/2)
print(z_neg_a_over_2)
print(z_a_over_2)



lover_bound = sam_mean + z_neg_a_over_2 * (sigma_X/(np.sqrt(100)))
upper_bound = sam_mean + z_a_over_2 * (sigma_X/(np.sqrt(100)))

SE = sigma_X/(np.sqrt(100)) # Standard Error

print(f"SE = {SE}")

#deviation
deviations = z_a_over_2 * SE
print(f"deviations = {deviations}")


print(f"lover_bound = {lover_bound}")
print(f"upper_bound = {np.round(upper_bound,2)}")

```

```python
from scipy.stats import norm
import numpy as np

#a)
print("a)")
a = 0.1

z_a_over_2 = norm.ppf(1-a/2)

print(f"z_a_over_2 = {z_a_over_2}")

#θ is unknown.   So we use the worst case:  θ=0.5


sigma_X = 0.5

n = (z_a_over_2 * sigma_X / 0.03) ** 2

print(f"n = {n}")
print(f"n = {np.ceil(n)}")



#b)
print("b)")

a = 0.01

z_a_over_2 = norm.ppf(1-a/2)

print(f"z_a_over_2 = {z_a_over_2}")

#θ is unknown.   So we use the worst case:  θ=0.5


sigma_X = 0.5

n = (z_a_over_2 * sigma_X / 0.03) ** 2

print(f"n = {n}")
print(f"n = {np.ceil(n)}")
```
