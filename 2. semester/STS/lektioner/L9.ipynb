Let $$X_1, X_2, X_3, X_4, X_5$$ be a random sample from a $$N(\mu,1)$$ distribution, where $$\mu$$ is unknown.

Suppose we observe
$$
5.45,\;4.23,\;7.22,\;6.94,\;5.98.
$$

We want to test
$$
H_0:\mu=\mu_0=5
\qquad \text{vs.} \qquad
H_1:\mu\neq 5.
$$

$$
\text{Define a test statistic to test the hypotheses and draw a conclusion assuming } \alpha = 0.05.
$$

```python
import numpy as np
from scipy.stats import norm
"""
Z = (obs_mu - mu) / (sigma / sqrt(n))

mu = 5
"""
mu = 5
var = 1
sigma = np.sqrt(var)
n = 5
obs = np.array([5.45, 4.23, 7.22, 6.94, 5.98])
obs_mu = np.mean(obs)

Z = (obs_mu - mu) / (sigma / np.sqrt(n))

"""
p = P(-a/2 < Z < a/2)
"""
c = norm.ppf(1-0.05/2)

print(f"Z: {Z}")
print(f"Critical value: {c}")

print(f"H0: mu = {mu}")
print(f"H1: mu != {mu}")

print(f"Conclusion: {'Reject H0' if abs(Z) > c else 'Fail to reject H0'}")
```

$$
\text{Find a }95\%\text{ confidence interval for }\bar{X}.
$$

$$
\text{Is }\mu_0\text{ included in the interval?}
$$

$$
\text{How does excluding }\mu_0\text{ from the interval relate to testing }H_0:\mu=\mu_0\text{ versus }H_1:\mu\neq\mu_0\text{?}
$$

```python
c = norm.ppf(1-0.05/2)
print(f"Critical value: {c}")

print(f"obs_mu: {obs_mu}")

"""
(-c < (obs_mu - mu) / (sigma / sqrt(n)) < c)
(-c*(sigma / sqrt(n) < (obs_mu - mu) < c*(sigma / sqrt(n))
(-obs_mu - c*(sigma / sqrt(n)) <  - mu < -obs_mu + c*(sigma / sqrt(n)))
(obs_mu - c*(sigma / sqrt(n)) <  mu < obs_mu + c*(sigma / sqrt(n)))
"""

print(f"Confidence interval: ({obs_mu - c*(sigma / np.sqrt(n))}, {obs_mu + c*(sigma / np.sqrt(n))})")
```

$$
\text{Let } X_1, X_2, \ldots, X_{150} \text{ be a random sample from an unknown distribution.}
$$

$$
\text{After observing this sample, the sample mean and sample variance are calculated as:}
$$
$$
\quad \bar{X} = 52.28,\; S^2 = 30.9.
$$

$$
\text{Design a level } \alpha = 0.05 \text{ test to choose between}
$$

$$
H_0:\mu = 50,
\qquad
H_1:\mu > 50.
$$

$$
\text{Do you accept or reject } H_0\text{?}
$$

```python
from scipy.stats import t
"""
Z = (obs_mu - mu) / (sigma / sqrt(n))

mu = 52.28
"""
mu = 50
S = 30.9
s = np.sqrt(S)
n = 150
obs_mu = 52.28

T = (obs_mu - mu) / (s / np.sqrt(n))

df = n - 1

"""
p = P(-a/2 < Z < a/2)
"""
a = 0.05
c = t.ppf(1-a, df)

print(f"T: {T}")
print(f"Critical value: {c}")

print(f"H0: mu = {mu}")
print(f"H1: mu > {mu}")

print(f"Conclusion: {'Reject H0' if T > c else 'Fail to reject H0'}")
```
